# 1. Introducción al lenguaje C  

El lenguaje C es **uno de los lenguajes de programación más importantes de la historia** y será el lenguaje principal de esta asignatura.

**Historia y origen:**
- Creado en **1972** por **Dennis Ritchie** en **Bell Labs**.  
- Inicialmente diseñado para **codificar el sistema operativo UNIX**, lo que lo hizo muy influyente en la informática moderna.

**Curiosidades interesantes:**
- Muchos lenguajes actuales, como **C++, C#, Java y Objective-C**, se basan en C.  
- La eficiencia y cercanía al hardware hacen que C siga siendo usado en:
  - **Sistemas embebidos** (microcontroladores, robots, IoT)  
  - **Sistemas operativos** (Linux, Windows, macOS)  
  - **Videojuegos y motores gráficos** (parte del motor de Unreal Engine, por ejemplo)  
- Su sintaxis minimalista permite **aprender los fundamentos de la programación**, como estructuras de control, funciones y manejo de memoria.

**Estándares de C:**
- **ANSI C (C89/C90)**: primera estandarización oficial.  
- **ISO C**: estandarización internacional.  
- **C99, C11, C18, C2x**: versiones modernas con mejoras, como soporte para tipos más grandes, funciones más seguras y bibliotecas extendidas.

> Aprender C no solo sirve para esta asignatura; te da **un pilar fundamental** para entender cómo funcionan los sistemas, el hardware y muchos otros lenguajes modernos.

### 2.1.1 Comentarios  

Los **comentarios** en C sirven para documentar el código:  
- Se escriben entre `/* ... */` o después de `//`  
- Mejoran la **lectura** y comprensión  
- Son **ignorados** por el compilador  
- Ayudan a otros programadores (o incluso a no programadores) a entender mejor el código  

<div style="background-color:#e6f3ff; padding:12px; border-radius:8px; width:60%; margin:auto;">

<pre><code>/*
  Nombre: Saludo.c
  Objetivo: Imprimir una línea de texto
*/

// Texto a imprimir: Hello world!!
</code></pre>

</div>



### 2.1.2 Directiva del preprocesador

Las **directivas del preprocesador** son instrucciones especiales que comienzan con el símbolo `#`.  

Se ejecutan **antes de la compilación** del programa y le indican al compilador ciertas tareas a realizar:   

- Las líneas que comienzan por `#` son procesadas por el **preprocesador**.  
- La directiva `#include <stdio.h>` indica que se debe **incluir el archivo de cabecera** estándar de **entrada/salida**.  
- Este archivo contiene las declaraciones necesarias para usar funciones como `printf()` y `scanf()`.  
- Existen otros archivos de cabecera que se pueden incluir (ej.: `<math.h>` para funciones matemáticas).  

<div style="background-color:#e6f3ff; padding:12px; border-radius:8px; width:60%; margin:auto;">

<pre><code><span style="color:#003366; font-weight:bold;">#include &lt;stdio.h&gt;</span>
</code></pre>

</div>


### 2.1.3 Función `main()`  

La función **`main()`** es fundamental en todo programa en C.  

- Forma parte de **todos los programas en C**.  
- Los paréntesis `()` después de `main` indican que se trata de una **función**.  
- Todo programa en C **comienza su ejecución** desde `main()`.  
- Toda función puede **retornar un valor** con `return`. En el caso de `main()`, siempre retorna un **entero (`int`)** al sistema operativo.  
- Una llave de apertura `{` marca el inicio del **bloque de instrucciones** de la función.  
- Una llave de cierre `}` marca el final de la función.  
- Es posible que `main()` reciba parámetros (`int main(int argc, char *argv[])`), pero en los ejemplos básicos se escribe vacío: `int main()`.  

<div style="background-color:#e6f3ff; padding:12px; border-radius:8px; width:60%; margin:auto;">

<pre><code><span style="color:#003366; font-weight:bold;">int main()</span> {
    return 0;
}
</code></pre>

</div>


> ⚠️ Nota: En ocasiones `main()` puede escribirse como  
> `int main(int argc, char *argv[])`.  
> Estos parámetros permiten **pasar información al programa** cuando se ejecuta.  
> Más adelante veremos cómo se utilizan, pero por ahora trabajaremos con la forma básica: `int main()`.


### 2.1.4 Sentencia `return`  

La sentencia **`return`** indica el valor que devuelve una función cuando finaliza.  
En el caso de `main()`, este valor se envía al **sistema operativo** como información sobre cómo terminó el programa.

- La función `main()` siempre retorna un **valor entero (`int`)**.  
- El valor `0` significa que el programa terminó **correctamente**.  
- Otros valores distintos de 0 suelen indicar que hubo algún **error o situación especial**.  
- El valor retornado no se muestra en pantalla, es solo **informativo para el sistema operativo**.  

<div style="background-color:#e6f3ff; padding:12px; border-radius:8px; width:60%; margin:auto;">

<pre><code>int main() {
    <span style="color:#003366; font-weight:bold;">return 0;</span>
}
</code></pre>

</div>




> Más adelante veremos ejemplos donde el valor de `return` puede usarse para **indicar errores** o diferentes resultados de ejecución.


### 2.1.5 Sentencia de salida  

La sentencia **`printf`** se utiliza para **imprimir información en pantalla**.  

- Imprime por pantalla la **cadena de caracteres** que aparece entre comillas.  
- La línea completa, incluyendo la función `printf`, su argumento entre paréntesis y el punto y coma (`;`), se llama **sentencia**.  

<div style="background-color:#e6f3ff; padding:12px; border-radius:8px; width:60%; margin:auto;">

<pre><code><span style="color:#003366; font-weight:bold;">printf("Hello world!!</span>\n<span style="color:#003366; font-weight:bold;">");</span>
</code></pre>

</div>


### 2.1.6 Carácter de nueva línea `\n`

En C, `\n` es un **carácter especial** que indica un **salto de línea** en la salida de pantalla.  

- Se llama **carácter de escape** porque no se imprime tal cual, sino que produce un efecto en la salida.  
- Permite que el texto que se imprime con `printf` continúe en la línea siguiente.  

<div style="background-color:#e6f3ff; padding:12px; border-radius:8px; width:60%; margin:auto;">

<pre><code>printf("Hello world!!<span style="color:#003366; font-weight:bold;">\n</span>");

## 2.2 Segundo paso: compilar  

Una vez que ya **reconocemos las partes de un programa en C**, el siguiente paso es **compilarlo**.  
La compilación transforma el código fuente (`.c`) en un **programa ejecutable** que la computadora puede ejecutar.

- En un terminal de Windows (CMD o PowerShell), se puede usar la siguiente orden:  

<div style="background-color:#e6f3ff; padding:12px; border-radius:8px; width:60%; margin:auto;">
<pre><code><span style="color:#003366; font-weight:bold;">> gcc saludo.c -o saludo</span>
</code></pre>
</div>

- Esta orden hace todo lo necesario para **generar el ejecutable `saludo`** a partir de un único archivo fuente.  
- No es necesario enlazar con otros archivos `.c`; con esta orden basta para obtener el programa ejecutable.  
- La compilación también **verifica errores de sintaxis** y prepara el código para ejecutarse.  

> **Nota:** Más adelante veremos un **esquema completo del proceso de compilación y ejecución**, que incluye preprocesado, compilación, enlazado y ejecución del programa.


## 2.3 Tercer paso: ejecutar

Una vez generado el archivo ejecutable, el programa ya está preparado para usarse.  
En Windows, se puede ejecutar desde el terminal con la siguiente orden:  

<div style="background-color:#e6f3ff; padding:12px; border-radius:8px; width:60%; margin:auto;">
<pre><code><span style="color:#003366; font-weight:bold;">> saludo</span>
</code></pre>
</div>

- También se puede pulsar la **opción correspondiente en el IDE** para ejecutar el programa.  
- Al ejecutar el programa, este **realiza las acciones definidas en la función `main()`**, como mostrar texto en pantalla mediante `printf`.  
- No es necesario ningún paso adicional si solo se usa un archivo fuente; el ejecutable generado con `-o` es suficiente para ejecutar el programa.  

> **Nota para Linux/macOS:** En estos sistemas sería necesario usar:  
> ```bash
> ./saludo
> ```
> Esto indica que el ejecutable se encuentra en el directorio actual.  

> **Nota general:** Más adelante veremos un **esquema completo del proceso de compilación y ejecución**, que incluye:  
> - Preprocesado  
> - Compilación  
> - Enlazado  
> - Generación del ejecutable  
> - Ejecución del programa

```bash
> Saludo2.exe
```  

> **Observaciones:**
>
> En Colab se requiere ./ porque Linux no busca en el directorio actual por defecto.  
>
> En Windows basta con el nombre del ejecutable.
>
> Con este ejercicio se visualiza claramente la secuencia de fases:
>
> <strong>Código fuente → Compilación → Ejecutable → Ejecución del programa.</strong>

<details>
  <summary>Solución (salida del programa). Clica aquí para ver la solución, pero piensa, incluso a partir del código, sin compilar y ejecutar el código dado.</summary>
  <pre>
Hello world!!
Welcome to C!
  </pre>
</details>
 


## 4.2 Ejemplo 2: contar líneas de salida

<div class="alert alert-info">

**EJERCICIO 2**:  

🧐 **Pregunta:** Observa el siguiente código y **responde cuántas líneas de texto se imprimirían en pantalla** si el programa se ejecutara.  

<div style="background-color:#e6f3ff; padding:12px; border-radius:8px; width:80%; margin:auto;">
<pre><code>/*
Nombre del archivo: Saludo2.c
Objetivo: Imprime texto
*/
#include <stdio.h>

int main ()
{
    <span style="color:#003366; font-weight:bold;">printf</span>("Hello world!!");
    <span style="color:#003366; font-weight:bold;">printf</span>("Welcome to C! ");
    return 0;
}
</code></pre>
</div>

<details>
<summary>Solución</summary>
<pre>
1
</pre>
</details>
 

# 5.  Entrada/Salida

## 5.1  Ejemplo 1: imprimiendo números

<div class="alert alert-info">

**EJERCICIO 3**:  

🧐 **Pregunta:** El siguiente programa <u>lee</u> un número de la <u>entrada estándar</u> y lo imprime dos veces por pantalla, pero los valores que muestra son distintos.  
¿Por qué ocurre esto?

<div style="background-color:#e6f3ff; padding:12px; border-radius:8px; width:80%; margin:auto;">
<pre><code><span style="color:#003366; font-weight:bold;">#include</span> &lt;stdio.h&gt;

<span style="color:#003366; font-weight:bold;">int</span> <span style="color:#003366; font-weight:bold;">main()</span>
{
    <span style="color:#003366; font-weight:bold;">int</span> number;
    <span style="color:#0055AA; font-weight:bold;">printf</span>("Introduce un numero: ");
    <span style="color:#800000; font-weight:bold;">scanf</span>("<span style="color:#003366; font-weight:bold;">%d</span>", &number);
    <span style="color:#0055AA; font-weight:bold;">printf</span>("Hexadecimal: <span style="color:#003366; font-weight:bold;">%X</span>\n Octal: <span style="color:#003366; font-weight:bold;">%o</span>\n", number, number);
    <span style="color:#003366; font-weight:bold;">return</span> 0;
}
</code></pre>
</div>

> Nota: Observa cómo se usa `printf` con **diferentes especificadores de formato** (`%X` y `%o`) para mostrar el mismo número de distinta forma. No es necesario ejecutar el programa para reflexionar sobre la salida.


<details>
<summary>Solución</summary>
<pre>
El programa muestra valores distintos porque se utiliza **diferentes formatos de impresión**: 
- `%X` muestra el número en **hexadecimal**. 
- `%o` muestra el número en **octal**. 
Aunque ambos corresponden al mismo valor interno, se representan de manera distinta en pantalla.
</pre>
</details>


### 5.1.1 Entrada/Salida  

A partir del programa del ejercicio 5.1, podemos comenzar a identificar conceptos básicos de **Entrada/Salida** en C:

- La <span style="color:#800000; font-weight:bold;">entrada</span> es el valor que el usuario teclea y que se almacena en la variable `number` mediante <span style="color:#800000; font-weight:bold;">scanf</span>.  
  Esto representa la **acción de leer datos desde el teclado**.

- La <span style="color:#0055AA; font-weight:bold;">salida</span> es la forma en que se muestra información por pantalla utilizando <span style="color:#0055AA; font-weight:bold;">printf</span>.  
  Esto representa la **acción de mostrar datos al usuario**.

> Nota: Este subapartado sirve solo como **introducción**. Más adelante veremos cómo manejar distintos tipos de datos, errores de entrada y formatos de salida más complejos en C.


## 5.2  Ejemplo 2: sumando números  

<div class="alert alert-info">

**EJERCICIO 4**:  

🧐 **Pregunta:** Escribe un programa en C que:

- <u>Lea</u> dos números reales desde el teclado (<span style="color:#800000; font-weight:bold;">entrada</span>) usando <span style="color:#800000; font-weight:bold;">scanf</span>.
- <u>Muestre</u> por pantalla (<span style="color:#0055AA; font-weight:bold;">salida</span>) la suma de los dos números usando <span style="color:#0055AA; font-weight:bold;">printf</span>.
</div>


### 5.2.1 Incluir operaciones en el programa  

Para ayudarte a plantear el programa, ten en cuenta algunas **pistas conceptuales**:

- Para leer los dos números reales que el usuario teclea, se puede usar la **entrada estándar** mediante <span style="color:#800000; font-weight:bold;">scanf</span>.  
  ⚠️ Los valores que el usuario ingresa **se almacenan en memoria**, y luego se pueden usar en operaciones.

- Para calcular la suma, se puede usar el operador <span style="color:#660066; font-weight:bold;">+</span>, y luego asignar el resultado a otra zona de memoria con el operador <span style="color:#660066; font-weight:bold;">=</span>.

- Para mostrar el resultado en pantalla, se puede usar la <span style="color:#0055AA; font-weight:bold;">salida</span> con <span style="color:#0055AA; font-weight:bold;">printf</span>.

- Flujo lógico que puedes tener en mente:  
  1. <span style="color:#800000; font-weight:bold;">Entrada</span> → memoria (con `scanf`)  
  2. Operación → memoria (con <span style="color:#660066; font-weight:bold;">+</span> y <span style="color:#660066; font-weight:bold;">=</span>)  
  3. <span style="color:#0055AA; font-weight:bold;">Salida</span> → pantalla (con `printf`)  


### 5.2.2 Código solución

A continuación se muestra una posible implementación del programa que lee dos números reales y muestra su suma:

<div style="background-color:#e6f3ff; padding:12px; border-radius:8px; width:80%; margin:auto;">
<pre><code><span style="color:#003366; font-weight:bold;">#include</span> &lt;stdio.h&gt;

<span style="color:#003366; font-weight:bold;">int</span> <span style="color:#003366; font-weight:bold;">main()</span>
{
    <span style="color:#003366; font-weight:bold;">float</span> number1, number2, output;
    <span style="color:#0055AA; font-weight:bold;">printf</span>("Enter the number: ");
    <span style="color:#800000; font-weight:bold;">scanf</span>("<span style="color:#003366; font-weight:bold;">%f</span>", &number1);
    <span style="color:#0055AA; font-weight:bold;">printf</span>("Enter another number: ");
    <span style="color:#800000; font-weight:bold;">scanf</span>("<span style="color:#003366; font-weight:bold;">%f</span>", &number2);
    <span style="color:#660066; font-weight:bold;">output</span> <span style="color:#660066; font-weight:bold;">=</span> number1 <span style="color:#660066; font-weight:bold;">+</span> number2;
    <span style="color:#0055AA; font-weight:bold;">printf</span>("The sum is: %f\n", output);
    <span style="color:#003366; font-weight:bold;">return</span> 0;
}
</code></pre>
</div>


### 5.2.3 Explicación de variables e identificadores  

En el programa anterior, las variables <span style="color:#003366; font-weight:bold;">number1</span>, <span style="color:#003366; font-weight:bold;">number2</span> y <span style="color:#003366; font-weight:bold;">output</span> son **espacios de memoria** donde se almacenan los datos que el usuario introduce y el resultado de la suma, respectivamente.

- Los nombres <span style="color:#003366; font-weight:bold;">number1</span>, <span style="color:#003366; font-weight:bold;">number2</span> y <span style="color:#003366; font-weight:bold;">output</span> se llaman **identificadores** en C.  
  * Se pueden crear tantos identificadores como se necesiten en un programa.  

- El tipo de dato <span style="color:#003366; font-weight:bold;">float</span> define que estas variables almacenan **números reales** (con decimales).  
  * Otro tipo podría ser <span style="color:#003366; font-weight:bold;">int</span>, que almacena **números enteros**.

> Nota: El tipo de dato de una variable determina cómo se almacena en memoria y qué operaciones se pueden realizar sobre ella.


<div class="alert alert-info" style="font-size:100%; width:80%; margin:auto;">

### Resumen de variables y operaciones

- Esta sentencia define tres espacios de memoria, es decir, tres variables:  
  <pre><code><span style="color:#003366; font-weight:bold;">float</span> number1, number2, output;</code></pre>
  
<br>

- Esta sentencia completa la operación suma, usando el operador <span style="color:#660066; font-weight:bold;">+</span>, y carga el resultado en la variable <span style="color:#003366; font-weight:bold;">output</span>:  
  <pre><code><span style="color:#003366; font-weight:bold;">output</span> <span style="color:#660066; font-weight:bold;">=</span> number1 <span style="color:#660066; font-weight:bold;">+</span> number2;</code></pre>  
  
<br>

> **Nota:** Este bloque sirve como recordatorio rápido de cómo se definen variables y se realizan operaciones básicas en C.

</div>


<div class="alert alert-info" style="font-size:100%">
<b>RECUERDA...</b>
    
**Arquitectura de Von Neumann: componentes y limitaciones**
    
- CPU central que ejecuta instrucciones y procesa datos.  
- Memoria compartida para instrucciones y datos.  
- Conexión a través de buses de datos, direcciones y control.  
- Dispositivos de entrada/salida que permiten la comunicación con el mundo exterior.  

📌 **Nota:** Este modelo presenta un **“cuello de botella de Von Neumann”**, ya que el bus compartido limita la velocidad a la que la CPU puede acceder simultáneamente a memoria e instrucciones. La mayoría de los ordenadores modernos añaden **memoria caché** para mitigar este efecto.

</div>


# 6. Escribe un programa

**Enunciado:**

- Es temporada de lanzamientos de videojuegos y una persona está esperando adquirir el último título de su franquicia favorita. Después de ir a la tienda y seleccionar el juego deseado, se dirige a la caja para efectuar la compra. Al llegar a la caja, se le aplica un descuento al precio del juego, como parte de una promoción especial.  

- Escriba un programa que calcule el **precio final de un videojuego aplicando un descuento**. El programa solicitará que se teclee el precio original del artículo y el descuento a aplicar, y mostrará el precio final del juego.

**ENTRADA:**  
- El precio del artículo.  
- El porcentaje de descuento.  

**SALIDA:**  
- El precio final a pagar.

> Se pide completar el siguiente código donde hay líneas incompletas:

```c
#include<stdio.h>

int main() {
    float initial_price, final_price;
    int percentage; // discount percentage

    printf("Enter the initial price: ");
    // Leer precio inicial sin descuento
    // ...

    printf("Enter the discount percentage: ");
    // Leer porcentaje de descuento
    // ...

    // Calcular precio final
    // ...

    printf("The final price is %g\n", final_price);
    return 0;
}
```

# 8. Y ahora con una función...

Este ejemplo es una **versión mejorada del programa del apartado 6**, usando una función para calcular el descuento.

- Una función permite **reutilizar código**, evitando repetir la misma lógica.
- Se define con un nombre, parámetros y valor de retorno.
- Se invoca desde `main()` para usar su lógica.

### Código completo

```c
#include <stdio.h>

// Declaración de la función
float calculateDiscount(float, int);

int main() {
    float initial_price, final_price;
    int percentage; // discount percentage

    printf("Enter the initial price: ");
    scanf("%f", &initial_price);

    printf("Enter the discount percentage: ");
    scanf("%d", &percentage);

    // INVOCACIÓN DE LA FUNCIÓN
    final_price = calculateDiscount(initial_price, percentage);

    printf("The final price is %g\n", final_price);
    return 0;
}

// DEFINICIÓN DE LA FUNCIÓN
float calculateDiscount(float initial_price, int percentage) {
    float final_price;
    final_price = initial_price - initial_price * percentage / 100;
    return final_price;
}
```

> **Observaciones:**
>
> - La línea `final_price = calculateDiscount(...)` muestra cómo se **invoca la función** desde `main()`.
> - La función `calculateDiscount` realiza todo el cálculo y **devuelve el resultado** al programa principal.
> - Más adelante se estudiarán funciones con **varios parámetros** y distintos tipos de retorno.
> - Usar funciones permite **organizar mejor el código** y facilita la **reutilización de la lógica**.


# 9. Añadiendo un bucle

A veces es necesario **repetir la misma acción varias veces**, por ejemplo, calcular precios con descuento para distintos juegos sin reiniciar el programa.

- Esto se puede implementar usando **bucles**: `for` o `while`.  
- Aquí vamos a usar un **bucle `while`** para repetir la entrada y el cálculo mientras el usuario lo desee.

### Código completo

```c
#include <stdio.h>

int main() {
    float initial_price, final_price;
    int percentage; // discount percentage
    int key = 1;

    while (key == 1) {  // <-- BUCLE WHILE
        printf("Enter the initial price: ");
        scanf("%f", &initial_price);

        printf("Enter the discount: ");
        scanf("%d", &percentage);

        final_price = initial_price - initial_price * percentage / 100;

        printf("The final price is %g\n", final_price);

        printf("Enter 1 to continue: ");   // mensaje para controlar el final del cálculo de descuentos
        scanf("%d", &key);                 // leer valor para repetir operación, o no
    }

    return 0;
}
```

> **Observaciones:**
> 
> - La línea del **bucle `while`** está remarcada en azul oscuro.  
> - El primer **`printf` antes del `scanf`** y el **`scanf` de entrada** están remarcados para que veas cómo se controla la repetición.  
> - Este bucle permite al usuario **repetir la acción tantas veces como quiera** hasta que introduzca otro valor distinto de 1.  
> - Más adelante veremos bucles `for` y otros tipos de sentencias de control en detalle.


# 10. Aplicando todo lo visto  

En el apartado anterior hemos visto cómo usar un bucle `while` para repetir la ejecución del programa.  
Ahora vamos a **combinarlo con una función**, de forma que el cálculo del precio con descuento se realice siempre en la misma función, y el bucle controle cuántas veces se quiere repetir el proceso.

### Código completo

```c
#include <stdio.h>

// Declaración de la función
float calculateDiscount(float, int);

int main() {
    float initial_price, final_price;
    int percentage;
    int key = 1;

    while (key == 1) {
        printf("Enter the initial price: ");
        scanf("%f", &initial_price);

        printf("Enter the discount percentage: ");
        scanf("%d", &percentage);

        // Invocación de la función
        final_price = calculateDiscount(initial_price, percentage);

        printf("The final price is %g\n", final_price);

        printf("Enter 1 to continue: ");
        scanf("%d", &key);
    }

    return 0;
}

// Definición de la función
float calculateDiscount(float initial_price, int percentage) {
    float final_price;
    final_price = initial_price - initial_price * percentage / 100;
    return final_price;
}


# 11. Actividades  

## 11.1 Actividad 1  

Escribe un programa que lea un número entero positivo y muestre por pantalla la secuencia de números de 0 hasta ese número.

**Entrada:**  
- Un número entero positivo.

**Salida:**  
- Secuencia de números de 0 hasta el número introducido.

<details>
<summary>💡 <strong>Pista</strong></summary>  

Piensa en una variable contador que empiece en 0 y vaya aumentando de 1 en 1, hasta llegar al número introducido.
</details>  

<br>
<details>
<summary><strong>Solución</strong></summary>

```c
#include <stdio.h>

int main() {
    int number, count = 0;

    printf("Enter a positive number: ");
    scanf("%d", &number);

    while (count <= number) {
        printf("%d\n", count);
        count = count + 1;   // incrementa el contador
    }

    return 0;
}
```

## 11.3 Repetir la secuencia 5 veces

Escribe un programa que repita **5 veces** la lectura de un número positivo y, para cada número, muestre la secuencia desde 0 hasta dicho valor.  

**Entrada**:  
- Un número positivo introducido por teclado, repetido 5 veces  
**Salida**:

<summary>💡 <strong>Pista</strong></summary>

1. leer un número

```c
#include <stdio.h>

// Declaración de la función
void count(int number);

int main() {
    int number;
    int times = 0;

    // repetir 5 veces
    while (times < 5) {
        printf("Enter a positive number: ");
        scanf("%d", &number);

        if (number >= 0) {
            count(number);  // llamada a la función
        } else {
            printf("Error: number must be positive.\n");
        }

        times = times + 1; // aumentar el contador
    }

    return 0;
}

// Definición de la función
void count(int number) {
    int i = 0;
    while (i <= number) {
        printf("%d ", i);
        i = i + 1;
    }
    printf("\n"); // salto de línea al final
}


# 12. 📚 Lecturas Recomendadas

Estas referencias complementan el contenido del notebook y ayudan a los alumnos a profundizar o contextualizar lo visto:

**1. Deitel, P. J., & Deitel, H. M. (2022)**  
**C: How to Program (10th ed.)**, Pearson.  
📘 *Referencia principal para ejemplos de programación y comprensión de cómo se representa la información en un ordenador. Útil para ver la relación entre números y caracteres.*

**2. Ceballos Sierra, F. J. (2019)**  
**C/C++. Curso de programación (5.ª ed.)**, RA-MA Editorial.  
📘 *Muy útil para repasar la sintaxis y estructuras de C, con ejemplos prácticos que complementan los conceptos vistos en el notebook.*

**3. Packt Publishing (2019)**  
**Learn C Programming**, Packt Publishing.  
📘 *Libro enfocado en aprender C paso a paso, con ejercicios prácticos y ejemplos claros sobre variables, operadores y estructuras de control.*


Otros recursos:

[**Compilador on-line**](https://www.onlinegdb.com/online_c_compiler)

> *© 2025, Julia Clemente Párraga (<mailto:juliaclementep@gmail.com>). Licensed under [MIT License](https://opensource.org/license/mit/)*
