# 8. Y ahora con una funci√≥n...

Este ejemplo es una **versi√≥n mejorada del programa del apartado 6**, usando una funci√≥n para calcular el descuento.

- Una funci√≥n permite **reutilizar c√≥digo**, evitando repetir la misma l√≥gica.
- Se define con un nombre, par√°metros y valor de retorno.
- Se invoca desde `main()` para usar su l√≥gica.

### C√≥digo completo

```c
#include <stdio.h>

// Declaraci√≥n de la funci√≥n
float calculateDiscount(float, int);

int main() {
    float initial_price, final_price;
    int percentage; // discount percentage

    printf("Enter the initial price: ");
    scanf("%f", &initial_price);

    printf("Enter the discount percentage: ");
    scanf("%d", &percentage);

    // INVOCACI√ìN DE LA FUNCI√ìN
    final_price = calculateDiscount(initial_price, percentage);

    printf("The final price is %g\n", final_price);
    return 0;
}

// DEFINICI√ìN DE LA FUNCI√ìN
float calculateDiscount(float initial_price, int percentage) {
    float final_price;
    final_price = initial_price - initial_price * percentage / 100;
    return final_price;
}
```

> **Observaciones:**
>
> - La l√≠nea `final_price = calculateDiscount(...)` muestra c√≥mo se **invoca la funci√≥n** desde `main()`.

# 9. A√±adiendo un bucle

A veces es necesario **repetir la misma acci√≥n varias veces**, por ejemplo, calcular precios con descuento para distintos juegos sin reiniciar el programa.

- Esto se puede implementar usando **bucles**: `for` o `while`.  
- Aqu√≠ vamos a usar un **bucle `while`** para repetir la entrada y el c√°lculo mientras el usuario lo desee.

### C√≥digo completo

```c
#include <stdio.h>

int main() {
    float initial_price, final_price;
    int percentage; // discount percentage
    int key = 1;

    while (key == 1) {  // <-- BUCLE WHILE
        printf("Enter the initial price: ");
        scanf("%f", &initial_price);

        printf("Enter the discount: ");
        scanf("%d", &percentage);

        final_price = initial_price - initial_price * percentage / 100;

        printf("The final price is %g\n", final_price);

        printf("Enter 1 to continue: ");   // mensaje para controlar el final del c√°lculo de descuentos
        scanf("%d", &key);                 // leer valor para repetir operaci√≥n, o no
    }

    return 0;
}
```

> **Observaciones:**
> 
> - La l√≠nea del **bucle `while`** est√° remarcada en azul oscuro.  
> - El primer **`printf` antes del `scanf`** y el **`scanf` de entrada** est√°n remarcados para que veas c√≥mo se controla la repetici√≥n.  
> - Este bucle permite al usuario **repetir la acci√≥n tantas veces como quiera** hasta que introduzca otro valor distinto de 1.  
> - M√°s adelante veremos bucles `for` y otros tipos de sentencias de control en detalle.


# 10. Aplicando todo lo visto  

En el apartado anterior hemos visto c√≥mo usar un bucle `while` para repetir la ejecuci√≥n del programa.  
Ahora vamos a **combinarlo con una funci√≥n**, de forma que el c√°lculo del precio con descuento se realice siempre en la misma funci√≥n, y el bucle controle cu√°ntas veces se quiere repetir el proceso.

### C√≥digo completo

```c
#include <stdio.h>

// Declaraci√≥n de la funci√≥n
float calculateDiscount(float, int);

int main() {
    float initial_price, final_price;
    int percentage;
    int key = 1;

    while (key == 1) {
        printf("Enter the initial price: ");
        scanf("%f", &initial_price);

        printf("Enter the discount percentage: ");
        scanf("%d", &percentage);

        // Invocaci√≥n de la funci√≥n
        final_price = calculateDiscount(initial_price, percentage);

        printf("The final price is %g\n", final_price);

        printf("Enter 1 to continue: ");
        scanf("%d", &key);
    }

    return 0;
}

// Definici√≥n de la funci√≥n
float calculateDiscount(float initial_price, int percentage) {
    float final_price;
    final_price = initial_price - initial_price * percentage / 100;
    return final_price;
}


# 11. Actividades  

## 11.1 Actividad 1  

Escribe un programa que lea un n√∫mero entero positivo y muestre por pantalla la secuencia de n√∫meros de 0 hasta ese n√∫mero.

**Entrada:**  
- Un n√∫mero entero positivo.

**Salida:**  
- Secuencia de n√∫meros de 0 hasta el n√∫mero introducido.

<details>
<summary>üí° <strong>Pista</strong></summary>  

Piensa en una variable contador que empiece en 0 y vaya aumentando de 1 en 1, hasta llegar al n√∫mero introducido.
</details>  

<br>
<details>
<summary><strong>Soluci√≥n</strong></summary>

```c
#include <stdio.h>

int main() {
    int number, count = 0;

    printf("Enter a positive number: ");
    scanf("%d", &number);

    while (count <= number) {
        printf("%d\n", count);
        count = count + 1;   // incrementa el contador
    }

    return 0;
}
```
