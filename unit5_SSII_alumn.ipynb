## 1. Introducción   

Cuando escribimos programas en C, es necesario controlar **el orden en el que se ejecutan las instrucciones**. A esto se le llama **control de flujo** y constituye la base de cualquier programa.

En programación, existen **tres formas fundamentales de control de flujo**:

1. **Secuencial**  
   - Es la forma más simple.  
   - Las instrucciones se ejecutan una detrás de otra, en el mismo orden en que fueron escritas en el código fuente.  
   - Ejemplo: leer un número, calcular su cuadrado y mostrarlo en pantalla.

2. **Selectiva (condicional)**  
   - Permite **elegir** si un bloque de código se ejecuta o no, dependiendo de que se cumpla una condición.  
   - En C se logra principalmente con las sentencias `if` y `switch`.

3. **Repetitiva (iterativa)**  
   - Sirve para **repetir** un conjunto de instrucciones varias veces, mientras se cumpla una condición o hasta que se alcance un número determinado de repeticiones.  
   - En C se utilizan las sentencias `while`, `do...while` y `for`.


A lo largo de este tema estudiaremos estas estructuras de control en C, con ejemplos prácticos que te permitirán aprender a escribir programas que **tomen decisiones, repitan procesos y controlen el flujo de ejecución de manera clara y ordenada**.


## 2. Sentencias de control selectivas

En la programación, muchas veces necesitamos que el programa **tome decisiones**: ejecutar un bloque de código en unas circunstancias y otro diferente en otras. Para esto se utilizan las **sentencias selectivas**, también llamadas **estructuras condicionales**.

En C, disponemos de varias formas de seleccionar qué instrucciones se ejecutan dependiendo de una condición o del valor de una expresión:

1. **`if`**  
   - Permite ejecutar un bloque de código *solo si* la condición que se evalúa es verdadera (`true`).  

2. **`if...else`**  
   - Ofrece dos alternativas:  
     - Se ejecuta un bloque de código si la condición es verdadera.  
     - Se ejecuta un bloque alternativo si la condición es falsa (`false`).  

3. **`switch`**  
   - Se utiliza cuando se necesita comparar una expresión contra múltiples valores posibles (llamados *casos*).  
   - Según el valor de la expresión, se ejecuta el bloque de código correspondiente.  

En los siguientes apartados veremos en detalle cada una de estas sentencias, con su sintaxis y ejemplos prácticos.


### 2.1 Sentencia `if`  

La sentencia `if` permite **ejecutar un bloque de código si una condición es verdadera**. Opcionalmente puede incluirse un bloque `else`, para indicar qué hacer en caso contrario.


#### Sintaxis general   

```c
if (condición)
    sentencia(s);         // o bloque de sentencias
[ else
    sentencia(s); ]       // o bloque de sentencias
```
> Los bloques de sentencias deben ir delimitados entre `{ }`, salvo que contengan una sola sentencia.

La condición es una **expresión numérica**:

- Se considera **verdadera** si el resultado es distinto de `0`.  
- Se considera **falsa** si el resultado es igual a `0`.

Estas expresiones pueden combinar los siguientes operadores (estudiados en el tema 3: elementos del lenguaje):

- **Operadores aritméticos**  
- **Operadores relacionales**  
- **Operadores lógicos** 

1. **Expresión numérica**

```
if (x)              // equivalente a if (x != 0)
    b /= x;         // se ejecuta solo si x es distinto de 0
x++;
```  

2. **Expresión relacional**

```
if (a < b)
    a++;
```

3. **Expresión lógica**

```
if (a < b && a > 0)
    a *= b;
``` 

#### 2.1.1 Sentencias `if` anidadas   

Una sentencia `if` puede incluir otra en su interior. Se debe tener cuidado con la indentación (sangría) y el uso de llaves, ya que esto afecta a qué bloque se asocia cada `else`.

### 2.2 Sentencia `switch`  

La sentencia `switch` permite seleccionar entre varias alternativas de ejecución en función del valor de una expresión entera.  
Es especialmente útil cuando hay que comparar una misma variable contra muchos valores posibles, evitando cadenas largas de `if` / `else if`.

#### Sintaxis general:

```c
switch (expresión) {
    case constante1:
        sentencia(s);
        [break;]
    [case constante2:
        sentencia(s);]
        [break;]
    ]
    ...
    [default:
        sentencia(s)_por_defecto;]
}
```

Calcular cuántos **días tiene un mes** de un año concreto.  
En el caso del mes de **febrero (2)** habrá que considerar si el año es **bisiesto**.

- **Entrada:** Dos enteros leídos desde teclado: mes (`mm`) y año (`aa`).  
- **Salida:** Indica cuántos días tiene el mes indicado, o informa si el mes no es válido.

<details>
<summary>Pista</summary>

- Lee `mm` y `aa` con `scanf`.  
- Usa `switch(mm)` y agrupa los `case` de los meses que comparten el mismo número de días (p. ej. 31 días).  
- En `case 2` (febrero) comprueba el bisiesto con la condición:
  `((aa % 4 == 0) && (aa % 100 != 0)) || (aa % 400 == 0)`.

- Imprime el resultado con `printf`.  
</details>

<details>
<summary>Solución</summary>

```c
#include <stdio.h>

int main() {
    int dd, mm, aa;

    // Pedimos mes y año al usuario
    printf("Introduce mes y año: ");
    scanf("%d %d", &mm, &aa);

    // Determinamos los días del mes
    switch (mm) {
        case 1: case 3: case 5: case 7: case 8: case 10: case 12:
            dd = 31;
            break;
        case 4: case 6: case 9: case 11:
            dd = 30;
            break;
        case 2:
            if ( ((aa % 4 == 0) && (aa % 100 != 0)) || (aa % 400 == 0) )
                dd = 29;
            else
                dd = 28;
            break;
        default:
            printf("El mes no es valido.\n");
    }

    // Solo imprimimos si el mes es válido
    if (mm >= 1 && mm <= 12)
        printf("El mes %d del anyo %d tiene %d dias.\n", mm, aa, dd);

    return 0;
}
```

## 3. Sentencias de control repetitivas   

En programación, muchas veces necesitamos **repetir un conjunto de instrucciones varias veces**, por ejemplo, leer varios datos, procesar elementos de una lista o ejecutar una acción hasta que se cumpla una condición. Para esto, C nos ofrece las **sentencias de repetición**, también llamadas **sentencias de iteración**.  

Estas sentencias permiten indicar al programa que **una acción se repita mientras una condición sea verdadera** (`true`) o hasta que se cumpla un criterio específico. Gracias a ellas podemos automatizar tareas que de otra forma requerirían escribir varias veces el mismo código, haciendo nuestros programas más eficientes y fáciles de mantener.  

En C existen tres estructuras principales de repetición:

- `while`: repite un bloque de código **mientras la condición sea verdadera**, comprobándola antes de ejecutar el bloque.
- `for`: ideal para iteraciones con un **número conocido de repeticiones**, combinando inicialización, condición y actualización en una sola línea.
- `do ... while`: similar a `while`, pero **ejecuta el bloque al menos una vez** antes de evaluar la condición.  

En los subapartados siguientes veremos **cada una de estas sentencias en detalle**, con ejemplos y buenas prácticas.



### 3.1 Sentencias `while` y `for` 

Las sentencias `while` y `for` permiten **repetir un bloque de código** mientras se cumpla una condición. Son estructuras equivalentes en cuanto a su capacidad de iteración, aunque su sintaxis y forma de expresar la repetición difieren ligeramente:

- `while` se usa cuando **no se sabe de antemano cuántas veces se repetirá** el bloque.
- `for` es ideal cuando **conocemos el número de iteraciones** o queremos controlar la variable de repetición de manera explícita.  

Ambas sentencias se pueden representar mediante un **diagrama común de flujo de control**, que muestra cómo se evalúa la condición y cuándo se ejecuta el bloque de código.

#### Sintaxis general de `while`:

```c
while (condición) {
    sentencia(s);
}
```

#### Sintaxis general de `for`:

```c
for (inicialización; condición; actualización) {
    sentencia(s);
}
```

### 3.2 Sentencia `do...while`  

La sentencia `do ... while` también permite **repetir un bloque de código mientras se cumpla una condición**, pero con una diferencia importante respecto a `while` y `for`:

- El bloque de código **siempre se ejecuta al menos una vez**, porque la condición se evalúa **después de ejecutar el bloque**.
- Es útil cuando necesitamos que el código se ejecute al menos una vez antes de comprobar la condición.

#### Sintaxis general de `do ... while`:

```c
do {
    sentencia(s);
} while (condición);
```

<h4 align="center"><strong>Observaciones</strong></h4>

<ul style="font-size:1.05rem;">
<li>En los tres bucles hay <b style="color:#1976d2;">inicialización</b>, <b style="color:#d32f2f;">condición</b> y <b style="color:#388e3c;">actualización</b>, aunque se ubiquen en distinto lugar.</li>
<li>La <b style="color:#d32f2f;">condición</b> es una <u>expresión lógica</u> que determina si el bucle continúa o se detiene.</li>
<li>La <b style="color:#388e3c;">actualización</b> es una <u>expresión aritmética o de asignación</u> que modifica el estado del bucle (por ejemplo, <code>i++</code>).</li>
<li><b><code>while</code></b> evalúa la condición <i>antes</i> de entrar en el bucle.</li>
<li><b><code>do ... while</code></b> la evalúa <i>después</i>, garantizando al menos una ejecución.</li>
<li><b><code>for</code></b> concentra todo en una línea, ideal para bucles con contador.</li>
</ul>

<h4 style="text-align:center;"><b>Diferencias entre <code>while</code> y <code>do ... while</code></b></h4>

> <b>do ... while</b>: el bloque de sentencias se ejecuta **al menos una vez**.<br>
> <b>while</b>: puede que el bloque de sentencias **no se ejecute ninguna vez**.

### 3.5 Bucles centinela y con contador

En programación, los **bucles** nos permiten repetir una serie de instrucciones. Dependiendo de cómo determinemos el número de iteraciones, podemos clasificarlos en **bucles centinela** y **bucles con contador**.


#### Bucle centinela

- El número de iteraciones **no está predeterminado**.  
- Depende de un **valor de entrada** que indica si el bucle continúa o termina.  

**Ejemplo: preguntar al usuario si quiere continuar**

```c
#include <stdio.h>

void limpia_buffer(void) {
    int c;
    // Vacía el buffer de entrada hasta fin de línea o fin de archivo
    while ((c = getchar()) != '\n' && c != EOF) {
        // no hace nada, solo limpia el resto de la línea
    }
}

int main(void) {
    int contador = 1;
    char respuesta;

    do {
        printf("%d\n", contador);
        contador++;

        printf("¿Quiere continuar? s:Sí, n:No: ");

        respuesta = getchar();   // lee el carácter introducido
        limpia_buffer();         // limpia el '\n' del buffer

    } while (respuesta == 's' || respuesta == 'S');

    printf("Programa terminado.\n");

    return 0;
}

```

#### Bucle con contador

- El número de iteraciones **está predeterminado**.  
- Se usa un **contador** que se incrementa hasta alcanzar un límite.  

**Ejemplo 1 – Bucle `while` extendido:**   

```c
#include <stdio.h>

int main() {
    int contador = 1;

    while (contador <= 10) {
        printf("%d\n", contador);
        contador++;
    }

    return 0;
}
```

## 4. Sentencias de salto en C (uso no recomendable)  

En C existen instrucciones que permiten **interrumpir o modificar el flujo normal de un bucle o condicional** de manera inmediata.  
Estas instrucciones se llaman **sentencias de salto**, y las más conocidas son `break` y `continue`.  

Aunque pueden ser útiles en ciertas situaciones, su uso **no es recomendable de manera general**, porque:

- Pueden hacer que el código sea **menos legible** y más difícil de mantener.  
- Rompen la secuencia normal de ejecución, lo que puede llevar a **errores sutiles** o inesperados.  
- En proyectos grandes, su uso excesivo puede dificultar la **comprensión y depuración** del programa.

**Cómo funcionan estas sentencias:**

- `break`: termina inmediatamente la ejecución del **bucle actual** y continúa con la instrucción siguiente al bucle.  
- `continue`: termina la **iteración actual** del bucle y pasa directamente a la **siguiente iteración** del mismo.  

En resumen, **ambas rompen el flujo normal del programa**, pero de manera diferente: `break` sale completamente del bucle, mientras que `continue` solo salta a la siguiente iteración.

Por eso, se recomienda aprender su funcionamiento, pero siempre **emplearlas con moderación** y solo cuando realmente simplifiquen el código.


### 4.1 Sentencia `break`  

La sentencia `break` permite **terminar inmediatamente un bucle**, independientemente de la condición del mismo.  
A continuación se muestran varios ejemplos prácticos con diferentes tipos de bucles.


#### Ejemplo 1:   

**`break` en un `while`**   

Con `break`:

```c
#include <stdio.h>

void limpia_buffer() {
    while (getchar() != '\n');
}

int main() {
    int i = 0;
    char car;

    while (i < 10) {
        printf("%d\n", i);
        printf("Continuar (n: no): ");
        
        if (getchar() == 'n')
            break;
        
        limpia_buffer();
        i++;
    }

    return 0;
}
```

**`while` sin `break` (condición incluida en el `while`)**   

```c
#include <stdio.h>

void limpia_buffer() {
    while (getchar() != '\n');
}

int main() {
    int i = 0;
    char car;

    while (i < 10 && getchar() != 'n') {
        printf("%d\n", i);
        printf("Continuar (n: no): ");
        car = getchar();
        limpia_buffer();
        i++;
    }

    return 0;
}
```

#### Ejemplo 2:   

**`for` con `break`**   

```c
#include <stdio.h>

void limpia_buffer() {
    while (getchar() != '\n');
}

int main() {
    int i;

    for (i = 0; i < 10; i++) {
        printf("%d\n", i);
        printf("Continuar? (n: no): ");
        
        if (getchar() == 'n')
            break;
        
        limpia_buffer();
    }

    return 0;
}
```

#### Ejemplo 3:   

**`do...while` con `break`**   

```c
#include <stdio.h>

void limpia_buffer() {
    while(getchar() != '\n');
}

int main() {
    int i = 0;

    do {
        printf("%d\n", i);
        printf("Continuar? (n: no): ");
        if (getchar() == 'n')
            break;
        limpia_buffer();
        i++;
    } while (i < 10);

    return 0;
}
```

💡 <strong>Recuerda:</strong><br><br>
La sentencia <code>break</code> rompe el flujo normal de ejecución del bucle, lo que puede dificultar la lectura y mantenimiento del código.<br><br>
Se recomienda evitar su uso excesivo y solo emplearla cuando realmente simplifique la lógica.
</div>


### 4.2 Sentencia `continue`

La sentencia `continue` permite **interrumpir la iteración actual de un bucle** y pasar directamente a la siguiente iteración.  
Su comportamiento depende del tipo de bucle:

- En `for`, el incremento se ejecuta automáticamente al final de cada iteración.  
- En `while` y `do...while`, si el incremento o actualización de la variable de control está después de `continue`, puede provocar **bucle infinito**.


#### Ejemplo 1: `continue` en un `for`

```c
#include <stdio.h>

int main() {
    int i;

    for (i = 0; i <= 10; i++) {
        if (i % 2)      // si i es impar
            continue;   // saltar a la siguiente iteración
        printf("%d\n", i);
    }

    return 0;
}
```
