## 1. Introducci√≥n   

Cuando escribimos programas en C, es necesario controlar **el orden en el que se ejecutan las instrucciones**. A esto se le llama **control de flujo** y constituye la base de cualquier programa.

En programaci√≥n, existen **tres formas fundamentales de control de flujo**:

1. **Secuencial**  
   - Es la forma m√°s simple.  
   - Las instrucciones se ejecutan una detr√°s de otra, en el mismo orden en que fueron escritas en el c√≥digo fuente.  
   - Ejemplo: leer un n√∫mero, calcular su cuadrado y mostrarlo en pantalla.

2. **Selectiva (condicional)**  
   - Permite **elegir** si un bloque de c√≥digo se ejecuta o no, dependiendo de que se cumpla una condici√≥n.  
   - En C se logra principalmente con las sentencias `if` y `switch`.

3. **Repetitiva (iterativa)**  
   - Sirve para **repetir** un conjunto de instrucciones varias veces, mientras se cumpla una condici√≥n o hasta que se alcance un n√∫mero determinado de repeticiones.  
   - En C se utilizan las sentencias `while`, `do...while` y `for`.


A lo largo de este tema estudiaremos estas estructuras de control en C, con ejemplos pr√°cticos que te permitir√°n aprender a escribir programas que **tomen decisiones, repitan procesos y controlen el flujo de ejecuci√≥n de manera clara y ordenada**.


## 2. Sentencias de control selectivas

En la programaci√≥n, muchas veces necesitamos que el programa **tome decisiones**: ejecutar un bloque de c√≥digo en unas circunstancias y otro diferente en otras. Para esto se utilizan las **sentencias selectivas**, tambi√©n llamadas **estructuras condicionales**.

En C, disponemos de varias formas de seleccionar qu√© instrucciones se ejecutan dependiendo de una condici√≥n o del valor de una expresi√≥n:

1. **`if`**  
   - Permite ejecutar un bloque de c√≥digo *solo si* la condici√≥n que se eval√∫a es verdadera (`true`).  

2. **`if...else`**  
   - Ofrece dos alternativas:  
     - Se ejecuta un bloque de c√≥digo si la condici√≥n es verdadera.  
     - Se ejecuta un bloque alternativo si la condici√≥n es falsa (`false`).  

3. **`switch`**  
   - Se utiliza cuando se necesita comparar una expresi√≥n contra m√∫ltiples valores posibles (llamados *casos*).  
   - Seg√∫n el valor de la expresi√≥n, se ejecuta el bloque de c√≥digo correspondiente.  

En los siguientes apartados veremos en detalle cada una de estas sentencias, con su sintaxis y ejemplos pr√°cticos.

## 3. Sentencias de control repetitivas   

En programaci√≥n, muchas veces necesitamos **repetir un conjunto de instrucciones varias veces**, por ejemplo, leer varios datos, procesar elementos de una lista o ejecutar una acci√≥n hasta que se cumpla una condici√≥n. Para esto, C nos ofrece las **sentencias de repetici√≥n**, tambi√©n llamadas **sentencias de iteraci√≥n**.  

Estas sentencias permiten indicar al programa que **una acci√≥n se repita mientras una condici√≥n sea verdadera** (`true`) o hasta que se cumpla un criterio espec√≠fico. En C existen tres estructuras principales de repetici√≥n:

- `while`: repite un bloque de c√≥digo **mientras la condici√≥n sea verdadera**, comprob√°ndola antes de ejecutar el bloque.
- `for`: ideal para iteraciones con un **n√∫mero conocido de repeticiones**, combinando inicializaci√≥n, condici√≥n y actualizaci√≥n en una sola l√≠nea.
- `do ... while`: similar a `while`, pero **ejecuta el bloque al menos una vez** antes de evaluar la condici√≥n.  

En los subapartados siguientes veremos **cada una de estas sentencias en detalle**, con ejemplos y buenas pr√°cticas.



### 3.1 Sentencias `while` y `for` 

Las sentencias `while` y `for` permiten **repetir un bloque de c√≥digo** mientras se cumpla una condici√≥n. Son estructuras equivalentes en cuanto a su capacidad de iteraci√≥n, aunque su sintaxis y forma de expresar la repetici√≥n difieren ligeramente:

- `while` se usa cuando **no se sabe de antemano cu√°ntas veces se repetir√°** el bloque.
- `for` es ideal cuando **conocemos el n√∫mero de iteraciones** o queremos controlar la variable de repetici√≥n de manera expl√≠cita.  

Ambas sentencias se pueden representar mediante un **diagrama com√∫n de flujo de control**, que muestra c√≥mo se eval√∫a la condici√≥n y cu√°ndo se ejecuta el bloque de c√≥digo.

#### Sintaxis general de `while`:

```c

```

#### Sintaxis general de `for`:

```c
for (inicializaci√≥n; condici√≥n; actualizaci√≥n) {
    sentencia(s);

```

### 3.5 Bucles centinela y con contador

En programaci√≥n, los **bucles** nos permiten repetir una serie de instrucciones. Dependiendo de c√≥mo determinemos el n√∫mero de iteraciones, podemos clasificarlos en **bucles centinela** y **bucles con contador**.


#### Bucle centinela

- El n√∫mero de iteraciones **no est√° predeterminado**.  
- Depende de un **valor de entrada** que indica si el bucle contin√∫a o termina.  

**Ejemplo: preguntar al usuario si quiere continuar**

```c
#include <stdio.h>

void limpia_buffer(void) {
    int c;
    // Vac√≠a el buffer de entrada hasta fin de l√≠nea o fin de archivo
    while ((c = getchar()) != '\n' && c != EOF) {
        // no hace nada, solo limpia el resto de la l√≠nea
    }
}

int main(void) {
    int contador = 1;
    char respuesta;

    do {
        printf("%d\n", contador);
        contador++;

        printf("¬øQuiere continuar? s:S√≠, n:No: ");

        respuesta = getchar();   // lee el car√°cter introducido
        limpia_buffer();         // limpia el '\n' del buffer

    } while (respuesta == 's' || respuesta == 'S');

    printf("Programa terminado.\n");

    return 0;
}

```

- El n√∫mero de iteraciones **est√° predeterminado**.

double power(long base, long exponent) {
    long pow, counter;
    
    if (exponent == 0)
        pow = 1;
    else {
        counter = 0;
        pow = 1;
        
        while (counter < labs(exponent)) { // labs => valor absoluto de long
            pow *= base;
            counter++;
        }
    }
    return exponent > 0 ? pow : 1/(double)pow;
}

## 4. Sentencias de salto en C (uso no recomendable)  

En C existen instrucciones que permiten **interrumpir o modificar el flujo normal de un bucle o condicional** de manera inmediata.  
Estas instrucciones se llaman **sentencias de salto**, y las m√°s conocidas son `break` y `continue`.  

Aunque pueden ser √∫tiles en ciertas situaciones, su uso **no es recomendable de manera general**, porque:

- Pueden hacer que el c√≥digo sea **menos legible** y m√°s dif√≠cil de mantener.  
- Rompen la secuencia normal de ejecuci√≥n, lo que puede llevar a **errores sutiles** o inesperados.  
- En proyectos grandes, su uso excesivo puede dificultar la **comprensi√≥n y depuraci√≥n** del programa.

**C√≥mo funcionan estas sentencias:**

- `break`: termina inmediatamente la ejecuci√≥n del **bucle actual** y contin√∫a con la instrucci√≥n siguiente al bucle.  
- `continue`: termina la **iteraci√≥n actual** del bucle y pasa directamente a la **siguiente iteraci√≥n** del mismo.  

En resumen, **ambas rompen el flujo normal del programa**, pero de manera diferente: `break` sale completamente del bucle, mientras que `continue` solo salta a la siguiente iteraci√≥n.

Por eso, se recomienda aprender su funcionamiento, pero siempre **emplearlas con moderaci√≥n** y solo cuando realmente simplifiquen el c√≥digo.


### 4.1 Sentencia `break`  

La sentencia `break` permite **terminar inmediatamente un bucle**, independientemente de la condici√≥n del mismo.  
A continuaci√≥n se muestran varios ejemplos pr√°cticos con diferentes tipos de bucles.

La sentencia `continue` permite **interrumpir la iteraci√≥n actual de un bucle** y pasar directamente a la siguiente iteraci√≥n.
