## 2. Matrices (arrays) unidimensionales

Un **array unidimensional** o **vector** es una colecci√≥n de elementos del mismo tipo, almacenados de forma contigua en memoria y accesibles mediante un **√≠ndice**.

```c
int a[10];       // Array de 10 enteros
double q[5];     // Array de 5 n√∫meros de tipo double
char c[20];      // Array de 20 caracteres
```

#### Ejemplo: Leer una serie de n√∫meros y mostrarlos.

> Fijarse en el uso de `MAX` para limitar la cantidad de elementos que se pueden almacenar.  

```c
#include <stdio.h>

#define MAX 8

int main() {
    int i, num;
    double A[MAX];

    // Pedir cantidad de n√∫meros
    printf("Cantidad de n√∫meros a leer (max. %d): ", MAX);
    scanf("%d", &num);

    // Validaci√≥n b√°sica
    if (num < 1) {
        printf("N√∫mero no v√°lido. Debe ser mayor o igual que 1.\n");
        return -1;
    } 
    else if (num > MAX) {
        printf("N√∫mero no v√°lido. No puede superar %d.\n", MAX);
        return -2;
    }

    // Lectura de los n√∫meros
    for (i = 0; i < num; i++) {
        printf("Elemento n√∫mero %d: ", i + 1);
        scanf("%lf", &A[i]);
    }

    // Mostrar los n√∫meros introducidos en la misma l√≠nea con tabulaciones y 3 decimales
    printf("\nLos n√∫meros introducidos son:\n");
    for (i = 0; i < num; i++)
        printf("\t%.3f", A[i]);  // tabulaci√≥n antes de cada n√∫mero y 3 decimales
    printf("\n");                 // salto de l√≠nea al final

    return 0;
}
```

### 3.2 Arrays como par√°metros a funciones  

<div style="border:2px solid #1E90FF; padding:10px; border-radius:5px; background-color:#E6F0FA;">
<b>Importante:</b> En una funci√≥n, los arrays se pasan como <b>par√°metros por referencia</b>.
</div>

Supongamos que tenemos los arrays:  

```c
double A[5] = {0};
double B[5] = {1,2,3,4,5};
```

Y definimos la funci√≥n:  

```c
int CopiarArrays(double X[], int num_x, double Y[], int num_y) {
    int i = 0;
    
    if (num_y > num_x) return 1; // Error: Y no cabe en X
    
    for (i = 0; i < num_y; i++)
        X[i] = Y[i];
    return num_y;
}
```

La llamada para copiar los arrays A y B ser√≠a:  

```c
int ret = 0;
int num_A = sizeof(A)/sizeof(double);
int num_B = sizeof(B)/sizeof(double);

ret = CopiarArrays(A, num_A, B, num_B);
``` 

**Representaci√≥n conceptual en memoria:**  

|   A   | A[0] | A[1] | A[2] | A[3] | A[4] |  ‚Ä¶  |   B   | B[0] | B[1] | B[2] | B[3] | B[4] |  ‚Ä¶  |
| :---: | :--: | :--: | :--: | :--: | :--: | :-: | :---: | :--: | :--: | :--: | :--: | :--: | :-: |
|       |   0  |   0  |   0  |   0  |   0  |  ‚Ä¶  |       |   1  |   2  |   3  |   4  |   5  |  ‚Ä¶  |
|   ‚Üì   |      |      |      |      |      |     |   ‚Üì   |      |      |      |      |      |     |
| &A[0] | A[0] | A[1] | A[2] | A[3] | A[4] |  ‚Ä¶  | &B[0] | B[0] | B[1] | B[2] | B[3] | B[4] |  ‚Ä¶  |


### `sizeof` de un array pasado como argumento a una funci√≥n   

<div style="border:2px solid #1E90FF; padding:10px; border-radius:5px; background-color:#E6F0FA;">
<b>Importante:</b> Cuando se pasa un array como par√°metro a una funci√≥n, <b>no se copia el array</b>. Lo que se pasa es la <b>direcci√≥n del primer elemento</b>. Por tanto, <b>sizeof dentro de la funci√≥n no da el tama√±o real del array</b>.
</div>

<br> 

Supongamos los arrays:

```c
double A[5] = {0};
double B[5] = {1,2,3,4,5};

int ret = 0;
int num_A = sizeof(A)/sizeof(double); // 5
int num_B = sizeof(B)/sizeof(double); // 5

ret = CopiarArrays(A, num_A, B, num_B);
```

Dentro de la funci√≥n:  

```c
int CopiarArrays(double X[], int num_x, double Y[], int num_y) {
    int i = 0;
    int n_x = sizeof(X)/sizeof(double); // ‚ùå NO es 5
    int n_y = sizeof(Y)/sizeof(double); // ‚ùå NO es 5

    // num_x y num_y contienen los tama√±os correctos
    for(i = 0; i < num_y; i++)
        X[i] = Y[i];
    return num_y;
}
``` 

> ‚ö†Ô∏è‚ö†Ô∏è‚ö†Ô∏è **IMPORTANTE:** Dentro de la funci√≥n se usan `num_x` y `num_y` para iterar, <span style="color:red;">NO</span> `sizeof(X)` ni `sizeof(Y)`.

### 3.3 Ejemplo: Mostrar una matriz   

```c
#include <stdio.h>
#include <stdlib.h>

void MostrarMatriz(int M[], int tam);

int main() {
    int A[] = {10, 20, 30, 40}; // 4 elementos
    int B[] = {1, 2, 3};        // 3 elementos

    printf("Array A: ");
    MostrarMatriz(A, sizeof(A)/sizeof(int)); // Pasamos el array y su tama√±o

    printf("Array B: ");
    MostrarMatriz(B, sizeof(B)/sizeof(int)); // Pasamos el array y su tama√±o

    return 0;
}

void MostrarMatriz(int M[], int tam) {
    int i;
    
    for (i = 0; i < tam; i++)
        printf("%d ", M[i]);
    printf("\n");
}
```

> F√≠jate c√≥mo se pasan arrays a funciones. No olvides que los arrays se **pasan por referencia** a funciones.  

### 3.4 Ejemplo: Asignar valores a una matriz   

```c
#include <stdio.h>
#include <stdlib.h>

void AsignarMatriz(int M[], int tam);

int main() {
    int A[4]; // Array de 4 elementos
    int B[3]; // Array de 3 elementos

    printf("Asignando valores al array A:\n");
    AsignarMatriz(A, sizeof(A)/sizeof(int));  // comenta aqu√≠ qu√© se est√° pasando a la funci√≥n como argumentos

    printf("Asignando valores al array B:\n");
    AsignarMatriz(B, sizeof(B)/sizeof(int)); 

    return 0;
}

void AsignarMatriz(int M[], int tam) {
    int i;
    
    for (i = 0; i < tam; i++) {
        printf("Elemento M[%d]: ", i);
        scanf("%d", &M[i]); // comenta aqu√≠ qu√© hace esta invocaci√≥n o llamada a scanf()
    }
    printf("\n");
}
```
> F√≠jate c√≥mo se pasan arrays a funciones. No olvides que los arrays se **pasan por referencia** a funciones.  

### 3.5 Indexaci√≥n  

En C, la indexaci√≥n de arrays es solo una forma m√°s legible de trabajar con punteros:  

```c
int A[5] = {10, 20, 30, 40, 50};
```

Podemos acceder a un elemento usando √≠ndices:  

```c
int x = A[2]; // x = 30
```

Pero tambi√©n podemos acceder a un elemento usando aritm√©tica de punteros:  

```c
int y = *(A + 2); // y = 30, equivalente a A[2]
```

> üòÖ Ya lo s√© de sobra, y m√°s en estos momentos, esto no te gusta tanto, pero al final ¬°es √∫til de alguna forma!


Y es que, la direcci√≥n de un elemento se puede expresar as√≠:  

```c
&A[2] == A + 2
```

Resumiendo:  

| Notaci√≥n |         Equivalencia con punteros         |
| :------: | :---------------------------------------: |
|  `A[i]`  |                 `*(A + i)`                |
|  `&A[i]` |                  `A + i`                  |
| `*(A+i)` | el contenido de la posici√≥n `i` del array |

### 4.5 Comparar cadenas de caracteres  

En C, la comparaci√≥n de cadenas se realiza mediante la funci√≥n `strcmp()`, incluida en la biblioteca `<string.h>`. Esta funci√≥n compara car√°cter a car√°cter dos cadenas de texto seg√∫n su orden alfab√©tico (lexicogr√°fico ASCII).


**Sintaxis:**  

```c
#include <string.h>

int strcmp(const char *cadena1, const char *cadena2);
```

**Par√°metros:**  

| Par√°metro | Descripci√≥n                                                  |
| :-------- | :----------------------------------------------------------- |
| `cadena1` | Puntero al primer array de caracteres que se va a comparar.  |
| `cadena2` | Puntero al segundo array de caracteres que se va a comparar. |


**Valor retornado:**   

| Valor devuelto | Significado                                                 |
| :------------- | :---------------------------------------------------------- |
| `0`            | Las cadenas son **iguales**.                                |
| `< 0`          | `cadena1` est√° **antes** en orden alfab√©tico que `cadena2`. |
| `> 0`          | `cadena2` est√° **antes** en orden alfab√©tico que `cadena1`. |

> ‚ö†Ô∏è La comparaci√≥n de `strcmp()` distingue may√∫sculas y min√∫sculas, ya que sigue el orden ASCII. Por ejemplo, "Hola" y "hola" no son iguales, porque 'H' (72) es diferente de 'h' (104).


**Ejemplo:**   

```c
#include <stdio.h>
#include <string.h>

int main(void) {
    char cadena1[20];
    char cadena2[20];
    int i;

    printf("Una frase: ");
    fgets(cadena1, sizeof(cadena1), stdin);

    printf("Otra frase: ");
    fgets(cadena2, sizeof(cadena2), stdin);

    i = strcmp(cadena1, cadena2);

    printf("\nOrdenadas:\n");

    if (i > 0) {
        puts(cadena2);
        puts(cadena1);
    } else if (i < 0) {
        puts(cadena1);
        puts(cadena2);
    } else {
        printf("Son iguales.\n");
    }

    return 0;
}
```

### 4.7 Funciones para conversi√≥n de datos  

En C, a veces necesitamos convertir datos almacenados como **cadenas de caracteres** a valores num√©ricos o modificar caracteres individuales. Para ello existen funciones de la biblioteca `<stdlib.h>` y `<ctype.h>`.

**Sintaxis:**

```c
#include <stdlib.h>

double atof(const char *cadena);  // convierte cadena a double

int atoi(const char *cadena);      // convierte cadena a int

long atol(const char *cadena);     // convierte cadena a long
```

```c
#include <ctype.h>
int tolower(int c);                // convierte un car√°cter a min√∫scula
int toupper(int c);                // convierte un car√°cter a may√∫scula
```

**Ejemplo 1:** Convertir cadenas a valores num√©ricos  

```c
#include <stdio.h>
#include <stdlib.h>
#include <string.h>

int main() {
    char s[50] = {0};
    double x = 0;
    int i = 0;
    long l = 0;

    strcpy(s, "3208.15E13");       // ejemplo atof
    x = atof(s);

    strcpy(s, "8.7195642337");     // otro ejemplo atof
    x = atof(s);

    strcpy(s, "887");               // ejemplo atoi
    i = atoi(s);

    strcpy(s, "89954");             // ejemplo atol
    l = atol(s);

    printf("x = %f, i = %d, l = %ld\n", x, i, l);

    return 0;
}

```

### 5.1 Asignar valores a elementos de un array multidimensional   

Podemos asignar valores a cada elemento de forma **individual**, ya sea directamente o mediante bucles `for`. Por claridad, vamos a centrarnos en un **array bidimensional**.

**Ejemplo: inicializaci√≥n con bucles anidados**

```c
#include <stdio.h>

#define FILAS    3
#define COLUMNAS 5

int main() {
    int A[FILAS][COLUMNAS];
    int i, j, n = 0;

    // Asignar valores secuenciales a cada elemento
    for(i = 0; i < FILAS; i++) {
        for(j = 0; j < COLUMNAS; j++, n++) {
            A[i][j] = n + 49; // ejemplo: asigna 49, 50, 51,...
        }
    }

    // Mostrar la matriz
    for(i = 0; i < FILAS; i++) {
        for(j = 0; j < COLUMNAS; j++) {
            printf("%d\t", A[i][j]);
        }
        printf("\n");
    }

    return 0;
}
```
