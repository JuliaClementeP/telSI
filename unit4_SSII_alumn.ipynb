# 2. Funciones y M√≥dulos: Perspectiva Actual

### Dise√±o *Top-Down*  

- El **dise√±o Top-Down** consiste en dividir un problema grande en partes m√°s peque√±as que se puedan resolver de forma separada.  
- Cada subproblema puede implementarse como:
  - **Funci√≥n:** un bloque de c√≥digo que realiza una tarea espec√≠fica dentro del programa.
  - **M√≥dulo:** un archivo o conjunto de funciones relacionadas que forman una unidad l√≥gica, facilitando la organizaci√≥n y la reutilizaci√≥n.
- En ejercicios iniciales, cada trozo se implementa normalmente como una **funci√≥n**; en programas m√°s grandes, los trozos pueden agruparse en **m√≥dulos** que contengan varias funciones relacionadas.
- Esta estrategia permite centrarse primero en la **visi√≥n general del problema** y luego en la **implementaci√≥n de cada pieza**, evitando perderse en detalles desde el inicio.


### Aplicaci√≥n en programaci√≥n

- Al aplicar Top-Down en programaci√≥n:
  - Se divide el problema en **funciones y m√≥dulos** y se centra el trabajo en el desarrollo de cada uno.  
  - Beneficios de la modularizaci√≥n:
    - Facilita entender la **naturaleza del problema**.  
    - Permite centrar la atenci√≥n en partes individuales antes de abordar el problema completo.  
    - Facilita la **prueba y depuraci√≥n** de cada m√≥dulo por separado.  
    - Promueve la **reutilizaci√≥n** de c√≥digo y la colaboraci√≥n en equipo.  

- **En un programa en C:**
  - Solo debe haber una **funci√≥n `main`**.  
  - La funci√≥n `main` puede usar:  
    - **Librer√≠as de funciones est√°ndar** (`stdio.h`, `math.h`)  
    - **Otras funciones definidas en el programa**.  

- Una funci√≥n en C consta de **dos partes principales**: la **declaraci√≥n** y la **definici√≥n**.  
- La **declaraci√≥n** indica al compilador que la funci√≥n **existir√°** y se podr√° usar en el programa, incluso antes de conocer su implementaci√≥n completa.  
- La **definici√≥n** es donde se escribe el c√≥digo que realiza realmente la tarea de la funci√≥n.  
- **Importante:** veremos la declaraci√≥n y definici√≥n de funciones en detalle m√°s adelante, cuando aprendamos a escribir funciones completas en C.   

### Buenas pr√°cticas modernas

- **Funciones cortas y espec√≠ficas:** cada funci√≥n realiza **una sola tarea**.  
- **M√≥dulos independientes:** cada m√≥dulo deber√≠a poder **compilarse y probarse por separado**.  
- **Documentaci√≥n clara:** comentar la finalidad de cada funci√≥n y los par√°metros que recibe.

```c
#include <stdio.h>

enum fruta { pera, limon, naranja };
typedef enum fruta Fruta;

int main() {
    Fruta sabor_del_zumo;

    printf("Introduce una fruta (0=pera, 1=limon, 2=naranja): ");
    scanf("%d", &sabor_del_zumo);

    printf("La fruta introducida es: %d\n", sabor_del_zumo);

    if (sabor_del_zumo == pera) printf("pera\n");
    else if (sabor_del_zumo == limon) printf("limon\n");
    else if (sabor_del_zumo == naranja) printf("naranja\n");

    return 0;
}



## 5. Agrupaci√≥n de instrucciones  

En C, las instrucciones pueden agruparse para organizar mejor el c√≥digo. Esto se hace mediante:

- **Sentencias simples**  
- **Bloques de sentencias**  
- **Funciones**  


### Sentencias simples

Una sentencia realiza una acci√≥n concreta:

```c
#include <stdio.h>

int main() {
    int a = 5, b = 3;
    int i;

    i = a + --b;  // sentencia simple
    printf("i = %d\n", i);

    return 0;
}
``` 

### Bloques de sentencias  

Un bloque agrupa varias sentencias entre `{ }`. Sirve para organizar c√≥digo y limitar el alcance de las variables.  

```c
#include <stdio.h>

int main() {
    int total = 100;

    {
        int descuento = 20;          // variable local al bloque
        int subtotal = total - descuento;
        printf("Subtotal despu√©s del descuento: %d\n", subtotal);
    }

    // Aqu√≠ no se puede usar descuento ni subtotal
    printf("Total original: %d\n", total);

    return 0;
}
``` 

### Funciones   

Una funci√≥n es un bloque de c√≥digo reutilizable asociado a un nombre. Puede recibir par√°metros y devolver un valor.  

<figure style="text-align: center;">
  <img src="figs/funcion_agrupacion.png" width="650"/>
  <figcaption>Fuente: Sistemas Inform√°ticos, UAH (2025)</figcaption>
</figure>


<div class="alert alert-success" style="font-family: monospace;">
üí° <strong>RECUERDA...</strong><br>
- Una funci√≥n puede aceptar <strong>cero o m√°s par√°metros</strong>.<br>
- Puede <strong>retornar un valor</strong> o ser <strong>void</strong> si no devuelve nada.<br>
- Permite <strong>reutilizar c√≥digo</strong> y mantener el programa organizado.
</div>


## 6. Declaraci√≥n y definici√≥n de una funci√≥n  

<div style="background:#fff9e6; padding:15px; border-radius:8px; font-family: monospace; font-size: 1.25em;">

#include &lt;stdio.h&gt;<br><br>

<b>int Porcentaje (int tot, int pc);</b>   <b><span style="color:#006400;">// DECLARACI√ìN o PROTOTIPO</span></b><br><br>

int main (void) {<br>
&nbsp;&nbsp;<b><span style="color:#006400;">// ...</span></b><br>
&nbsp;&nbsp;iva = <b>Porcentaje(</b>total, porcen<b>)</b>;   <b><span style="color:#006400;">// LLAMADA o EJECUCI√ìN</span></b><br>
&nbsp;&nbsp;<b><span style="color:#006400;">// ...</span></b><br>
}<br><br>

<b>int Porcentaje (int tot, int pc)</b><br>
<b>{</b>   <b><span style="color:#006400;">// DEFINICI√ìN o IMPLEMENTACI√ìN</span></b><br>
&nbsp;&nbsp;<b><span style="color:#006400;">// ...</span></b><br>
<b>}</b><br>

</div>




## 7. Paso de par√°metros a una funci√≥n   


En C, las funciones pueden recibir **informaci√≥n desde el programa principal** u otras funciones mediante **par√°metros**.  
Los par√°metros permiten que una funci√≥n **trabaje con datos distintos cada vez que se llama**, aumentando la **reutilizaci√≥n del c√≥digo** y la **flexibilidad**.

Cuando se pasa un par√°metro a una funci√≥n, hay dos formas principales de hacerlo:

1. **Paso por valor:** la funci√≥n recibe una copia del dato.  
2. **Paso por referencia:** la funci√≥n recibe la direcci√≥n del dato, pudiendo modificar la variable original.

> üí° Usar correctamente el tipo de paso de par√°metros es fundamental para controlar **qu√© se modifica y qu√© no**, as√≠ como para optimizar memoria y eficiencia.


### 7.1 Paso de par√°metros por valor  

<div style="background:#fff9e6; padding:15px; border-radius:8px; font-family: monospace; font-size: 1.25em;">

#include &lt;stdio.h&gt;<br><br>

<b>void intercambio(int x, int y);</b>   <b><span style="color:#006400;">// DECLARACI√ìN o PROTOTIPO</span></b><br><br>

<b>int main(void)</b>   <b><span style="color:#006400;">// NUNCA usar void main()</span></b><br>
<b>{</b><br>
&nbsp;&nbsp;int a = 20;<br>
&nbsp;&nbsp;int b = 30;<br>
&nbsp;&nbsp;<b><span style="color:#006400;">// Mostramos valores y direcciones antes de la funci√≥n</span></b><br>
&nbsp;&nbsp;printf("Main: a=%d (%p)\n", a, &a);<br>
&nbsp;&nbsp;printf("Main: b=%d (%p)\n", b, &b);<br>
<br>
&nbsp;&nbsp;<b><span style="background:#bbdefb; color:#000000;">// CLAVE: a ‚Üí x, b ‚Üí y (paso por valor)</span></b><br>
&nbsp;&nbsp;intercambio(a, b);<br>
<br>
&nbsp;&nbsp;<b><span style="color:#006400;">// Mostramos valores y direcciones despu√©s de la funci√≥n</span></b><br>
&nbsp;&nbsp;printf("Main despu√©s: a=%d (%p)\n", a, &a);<br>
&nbsp;&nbsp;printf("Main despu√©s: b=%d (%p)\n", b, &b);<br>
&nbsp;&nbsp;return 0;<br>
<b>}</b><br><br>

<b>void intercambio(int x, int y)</b><br>
<b>{</b>   <b><span style="color:#006400;">// DEFINICI√ìN o IMPLEMENTACI√ìN</span></b><br>
<br>
&nbsp;&nbsp;<b><span style="background:#bbdefb; color:#000000;">// CLAVE: x y y reciben copias de los valores de a y b (a ‚Üí x, b ‚Üí y)</span></b><br>
<br>
&nbsp;&nbsp;<b><span style="color:#006400;">// Mostramos valores y direcciones al inicio</span></b><br>
&nbsp;&nbsp;printf("Antes swap: x=%d (%p)\n", x, &x);<br>
&nbsp;&nbsp;printf("Antes swap: y=%d (%p)\n", y, &y);<br>
<br>
&nbsp;&nbsp;<b><span style="color:#006400;">// Intercambiamos valores</span></b><br>
&nbsp;&nbsp;int temp;<br>
&nbsp;&nbsp;temp = x;<br>
&nbsp;&nbsp;x = y;<br>
&nbsp;&nbsp;y = temp;<br>
<br>
&nbsp;&nbsp;<b><span style="color:#006400;">// Mostramos valores y direcciones despu√©s del swap</span></b><br>
&nbsp;&nbsp;printf("Despu√©s swap: x=%d (%p)\n", x, &x);<br>
&nbsp;&nbsp;printf("Despu√©s swap: y=%d (%p)\n", y, &y);<br>
<b>}</b><br>

</div>

<div class="alert alert-info" style="font-family: monospace; font-size:1.1em;">
üí° <strong>Nota:</strong> ¬øFunciona bien? Tecl√©alo y comprueba.
</div>



### **Visualizaci√≥n de memoria paso por valor:**   

<div style="font-family: monospace; font-size:1.25em;">

<h4>Antes de la llamada</h4>
<table style="border-collapse: collapse;">
<tr>
  <th style="border:1px solid black; padding:5px;">Variable</th>
  <th style="border:1px solid black; padding:5px;">Direcci√≥n</th>
  <th style="border:1px solid black; padding:5px;">Valor</th>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">a</td>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">0x7ffee1f4</td>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">20</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">b</td>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">0x7ffee1f0</td>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">30</td>
</tr>
</table>

<h4>Dentro de intercambio (inicio)</h4>
<table style="border-collapse: collapse;">
<tr>
  <th style="border:1px solid black; padding:5px;">Variable</th>
  <th style="border:1px solid black; padding:5px;">Direcci√≥n</th>
  <th style="border:1px solid black; padding:5px;">Valor</th>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#b9f6ca;">x</td>
  <td style="border:1px solid black; padding:5px; background:#b9f6ca;">0x7fffdf10</td>
  <td style="border:1px solid black; padding:5px; background:#b9f6ca;">20</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#69f0ae;">y</td>
  <td style="border:1px solid black; padding:5px; background:#69f0ae;">0x7fffdf0c</td>
  <td style="border:1px solid black; padding:5px; background:#69f0ae;">30</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#00c853;">temp</td>
  <td style="border:1px solid black; padding:5px; background:#00c853;">0x7fffdf08</td>
  <td style="border:1px solid black; padding:5px; background:#00c853;">-</td>
</tr>
</table>

<h4>Dentro de intercambio (despu√©s swap)</h4>
<table style="border-collapse: collapse;">
<tr>
  <th style="border:1px solid black; padding:5px;">Variable</th>
  <th style="border:1px solid black; padding:5px;">Direcci√≥n</th>
  <th style="border:1px solid black; padding:5px;">Valor</th>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#b9f6ca;">x</td>
  <td style="border:1px solid black; padding:5px; background:#b9f6ca;">0x7fffdf10</td>
  <td style="border:1px solid black; padding:5px; background:#b9f6ca;">30</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#69f0ae;">y</td>
  <td style="border:1px solid black; padding:5px; background:#69f0ae;">0x7fffdf0c</td>
  <td style="border:1px solid black; padding:5px; background:#69f0ae;">20</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#00c853;">temp</td>
  <td style="border:1px solid black; padding:5px; background:#00c853;">0x7fffdf08</td>
  <td style="border:1px solid black; padding:5px; background:#00c853;">20</td>
</tr>
</table>

<h4>Despu√©s de volver a main</h4>
<table style="border-collapse: collapse;">
<tr>
  <th style="border:1px solid black; padding:5px;">Variable</th>
  <th style="border:1px solid black; padding:5px;">Direcci√≥n</th>
  <th style="border:1px solid black; padding:5px;">Valor</th>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">a</td>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">0x7ffee1f4</td>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">20</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">b</td>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">0x7ffee1f0</td>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">30</td>
</tr>
</table>

</div>



<div class="alert alert-info" style="font-family: monospace; font-size:1.1em;">
üí° <strong>Nota:</strong> Se muestran solo las celdas implicadas en este ejemplo. Cada variable tiene su propia direcci√≥n, y puede imaginarse que est√°n consecutivas en la <strong>pila</strong> (√°rea de memoria donde se almacenan las variables locales de cada funci√≥n y la direcci√≥n a la que debe volver la ejecuci√≥n). Esto permite entender c√≥mo funciona el <strong>paso por valor</strong>: las copias locales de la funci√≥n no afectan a las variables originales en <strong>main</strong>.
</div>


### 7.2 Paso de par√°metros por referencia   

<div style="background:#fff9e6; padding:15px; border-radius:8px; font-family: monospace; font-size: 1.25em;">

#include &lt;stdio.h&gt;<br><br>

<b>void intercambio(int *p, int *q);</b>   <b><span style="color:#006400;">// DECLARACI√ìN o PROTOTIPO</span></b><br><br>

<b>int main(void)</b>   <b><span style="color:#006400;">// NUNCA usar void main()</span></b><br>
<b>{</b><br>
&nbsp;&nbsp;int a = 20;<br>
&nbsp;&nbsp;int b = 30;<br>
&nbsp;&nbsp;<b><span style="color:#006400;">// Mostramos valores y direcciones antes de la funci√≥n</span></b><br>
&nbsp;&nbsp;printf("Main: a=%d (%p)\n", a, &a);<br>
&nbsp;&nbsp;printf("Main: b=%d (%p)\n", b, &b);<br>
<br>
&nbsp;&nbsp;<b><span style="background:#bbdefb; color:#000000;">// CLAVE: pasamos las direcciones de a y b (a ‚Üí *p, b ‚Üí *q)</span></b><br>
&nbsp;&nbsp;intercambio(&a, &b);<br>
<br>
&nbsp;&nbsp;<b><span style="color:#006400;">// Mostramos valores y direcciones despu√©s de la funci√≥n</span></b><br>
&nbsp;&nbsp;printf("Main despu√©s: a=%d (%p)\n", a, &a);<br>
&nbsp;&nbsp;printf("Main despu√©s: b=%d (%p)\n", b, &b);<br>
&nbsp;&nbsp;return 0;<br>
<b>}</b><br><br>

<b>void intercambio(int *p, int *q)</b><br>
<b>{</b>   <b><span style="color:#006400;">// DEFINICI√ìN o IMPLEMENTACI√ìN</span></b><br>
<br>
&nbsp;&nbsp;<b><span style="background:#bbdefb; color:#000000;">// CLAVE: p apunta a a, q apunta a b (paso por referencia)</span></b><br>
<br>
&nbsp;&nbsp;<b><span style="color:#006400;">// Mostramos valores y direcciones al inicio</span></b><br>
&nbsp;&nbsp;printf("Antes swap: *p=%d (%p)\n", *p, p);<br>
&nbsp;&nbsp;printf("Antes swap: *q=%d (%p)\n", *q, q);<br>
&nbsp;&nbsp;<b><span style="color:#006400;">// Intercambiamos valores</span></b><br>
&nbsp;&nbsp;int temp;<br>
&nbsp;&nbsp;temp = *p;<br>
&nbsp;&nbsp;*p = *q;<br>
&nbsp;&nbsp;*q = temp;<br>
&nbsp;&nbsp;<b><span style="color:#006400;">// Mostramos valores y direcciones despu√©s del swap</span></b><br>
&nbsp;&nbsp;printf("Despu√©s swap: *p=%d (%p)\n", *p, p);<br>
&nbsp;&nbsp;printf("Despu√©s swap: *q=%d (%p)\n", *q, q);<br>
<b>}</b><br>

</div>


### **Visualizaci√≥n de memoria paso por referencia:**   

<div style="font-family: monospace; font-size:1.25em;">

<h4>Antes de la llamada</h4>
<table style="border-collapse: collapse;">
<tr>
  <th style="border:1px solid black; padding:5px;">Variable</th>
  <th style="border:1px solid black; padding:5px;">Direcci√≥n</th>
  <th style="border:1px solid black; padding:5px;">Contenido / Valor</th>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">a</td>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">0x7ffee1f4</td>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">20</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">b</td>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">0x7ffee1f0</td>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">30</td>
</tr>
</table>

<h4>Dentro de intercambio (inicio)</h4>
<table style="border-collapse: collapse;">
<tr>
  <th style="border:1px solid black; padding:5px;">Variable</th>
  <th style="border:1px solid black; padding:5px;">Direcci√≥n</th>
  <th style="border:1px solid black; padding:5px;">Contenido / Valor</th>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">a</td>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">0x7ffee1f4</td>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">20</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">b</td>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">0x7ffee1f0</td>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">30</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#b9f6ca;">p</td>
  <td style="border:1px solid black; padding:5px; background:#b9f6ca;">0x7fffdf10</td>
  <td style="border:1px solid black; padding:5px; background:#b9f6ca;">0x7ffee1f4</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#69f0ae;">q</td>
  <td style="border:1px solid black; padding:5px; background:#69f0ae;">0x7fffdf0c</td>
  <td style="border:1px solid black; padding:5px; background:#69f0ae;">0x7ffee1f0</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#00c853;">temp</td>
  <td style="border:1px solid black; padding:5px; background:#00c853;">0x7fffdf08</td>
  <td style="border:1px solid black; padding:5px; background:#00c853;">-</td>
</tr>
</table>

<h4>Dentro de intercambio (despu√©s swap)</h4>
<table style="border-collapse: collapse;">
<tr>
  <th style="border:1px solid black; padding:5px;">Variable</th>
  <th style="border:1px solid black; padding:5px;">Direcci√≥n</th>
  <th style="border:1px solid black; padding:5px;">Contenido / Valor</th>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">a</td>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">0x7ffee1f4</td>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">30</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">b</td>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">0x7ffee1f0</td>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">20</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#b9f6ca;">p</td>
  <td style="border:1px solid black; padding:5px; background:#b9f6ca;">0x7fffdf10</td>
  <td style="border:1px solid black; padding:5px; background:#b9f6ca;">0x7ffee1f4</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#69f0ae;">q</td>
  <td style="border:1px solid black; padding:5px; background:#69f0ae;">0x7fffdf0c</td>
  <td style="border:1px solid black; padding:5px; background:#69f0ae;">0x7ffee1f0</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#00c853;">temp</td>
  <td style="border:1px solid black; padding:5px; background:#00c853;">0x7fffdf08</td>
  <td style="border:1px solid black; padding:5px; background:#00c853;">30</td>
</tr>
</table>

<h4>Despu√©s de volver a main</h4>
<table style="border-collapse: collapse;">
<tr>
  <th style="border:1px solid black; padding:5px;">Variable</th>
  <th style="border:1px solid black; padding:5px;">Direcci√≥n</th>
  <th style="border:1px solid black; padding:5px;">Contenido / Valor</th>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">a</td>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">0x7ffee1f4</td>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">30</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">b</td>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">0x7ffee1f0</td>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">20</td>
</tr>
</table>

</div>

<div class="alert alert-info" style="font-family: monospace; font-size:1.1em;">
üí° <strong>Nota:</strong> Se muestran solo las celdas implicadas en este ejemplo. Cada variable tiene su propia direcci√≥n, y puede imaginarse que est√°n consecutivas en la <strong>pila</strong>. Los punteros <strong>p</strong> y <strong>q</strong> contienen las direcciones de <strong>a</strong> y <strong>b</strong>, y al modificarlos mediante <strong>*p</strong> y <strong>*q</strong>, los valores originales cambian en <strong>main</strong>.
</div>


## 8. Programa formado por varios archivos  

Cuando un programa crece, conviene **separar las funciones en varios archivos** para organizar mejor el c√≥digo y facilitar su mantenimiento.

### Archivo 1: `modulo01.c` (funci√≥n `main`)

```c
// modulo01.c: funci√≥n main
#include <stdio.h>

// Declaraci√≥n de la funci√≥n Media
double Media(double, double);

int main(void)  // nunca void main()
{
    double x = 2.34;
    double y = 1.73;
    double m;

    // Llamada a la funci√≥n Media
    m = Media(x, y);

    printf("Media == %f\n", m);

    return 0;
}
```

## 10. Visibilidad de variables

En C, la **visibilidad** de una variable depende de d√≥nde se declare:

<div class="alert alert-primary" style="font-family: monospace; font-size:1.1em;">
üü¶ <strong>Variables globales</strong><br>
- Se declaran fuera de cualquier funci√≥n.<br>
- Son visibles desde su definici√≥n en adelante, en todas las funciones del archivo.<br>
- √ötiles para informaci√≥n compartida.
</div>

<div class="alert alert-success" style="font-family: monospace; font-size:1.1em;">
üü© <strong>Variables locales</strong><br>
- Se declaran dentro de un bloque (por ejemplo, dentro de `main()` o un `if`).<br>
- Solo son visibles dentro de ese bloque y sus sub-bloques.
</div>

<div class="alert alert-warning" style="font-family: monospace; font-size:1.1em;">
üü® <strong>Variables con nombres iguales</strong><br>
- Pueden existir en bloques diferentes.<br>
- Si un bloque es sub-bloque de otro, la variable del sub-bloque <strong>eclipsa</strong> a la del bloque superior.<br>
- El tipo puede ser distinto, no importa.
</div>


### Ejemplo sencillo:

```c
#include <stdio.h>

// Variable global
int global = 10;

int main(void)
{
    int local = 5; // Variable local en main

    printf("Global: %d, Local: %d\n", global, local);

    {   // Inicio de un sub-bloque
        int local = 20; // Eclipsa a la variable 'local' de main
        int sub = 7;
        printf("Dentro sub-bloque -> Local eclipsada: %d, Sub: %d\n", local, sub);
    }

    // Aqu√≠ 'sub' ya no existe, pero 'local' de main sigue visible
    printf("Despu√©s del sub-bloque -> Local: %d\n", local);

    return 0;
}


## 11. Calificaci√≥n de variables

En C, los calificadores modifican el **comportamiento de las variables**, afectando su **duraci√≥n, visibilidad y forma de acceso**.  

### Tabla de calificaci√≥n

<table style="border-collapse: collapse; font-family: monospace; font-size:1.1em;">
<tr>
  <th style="border:1px solid black; padding:5px;">Calificador</th>
  <th style="border:1px solid black; padding:5px;">Local</th>
  <th style="border:1px solid black; padding:5px;">Global</th>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px;">auto (por defecto)</td>
  <td style="border:1px solid black; padding:5px;">Se crea, inicializa y destruye cada vez que se ejecuta el bloque.</td>
  <td style="border:1px solid black; padding:5px;">No aplicable</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px;">static</td>
  <td style="border:1px solid black; padding:5px;">Permanece entre llamadas; inicializa solo la primera vez.</td>
  <td style="border:1px solid black; padding:5px;">Accesible solo desde el archivo donde se define.</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px;">extern</td>
  <td style="border:1px solid black; padding:5px;">No aplicable</td>
  <td style="border:1px solid black; padding:5px;">Indica que la variable global est√° definida en otro archivo o en otro lugar del mismo archivo.</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px;">volatile</td>
  <td style="border:1px solid black; padding:5px;">No almacenar en registros ni en cach√© del compilador; cada lectura/escritura va directamente a memoria principal.</td>
  <td style="border:1px solid black; padding:5px;">Igual que local: no almacenar en registros; siempre leer/escribir en memoria principal.</td>
</tr>
</table>

### Ejemplo pr√°ctico

```c
#include <stdio.h>

int global = 5;          // variable global normal
static int archivo = 10; // variable global est√°tica

int main(void)
{
    auto int local = 3;      // auto, por defecto
    static int contador = 0; // local est√°tica, persiste entre llamadas
    volatile int sensor = 7; // local volatile

    contador++;
    printf("Global: %d, Local: %d, Contador: %d, Sensor: %d\n",
           global, local, contador, sensor);

    return 0;
}


## 12. üìö Lecturas Recomendadas

Estas referencias complementan el contenido del notebook y ayudan a los alumnos a profundizar o contextualizar lo visto:

**1. Deitel, P. J., & Deitel, H. M. (2022)**  
**C: How to Program (10th ed.)**, Pearson.  
üìò *Referencia principal para ejemplos de programaci√≥n y comprensi√≥n de c√≥mo se representa la informaci√≥n en un ordenador. √ötil para ver la relaci√≥n entre n√∫meros y caracteres.*

**2. Ceballos Sierra, F. J. (2019)**  
**C/C++. Curso de programaci√≥n (5.¬™ ed.)**, RA-MA Editorial.  
üìò *Muy √∫til para repasar la sintaxis y estructuras de C, con ejemplos pr√°cticos que complementan los conceptos vistos en el notebook.*

**3. Packt Publishing (2019)**  
**Learn C Programming**, Packt Publishing.  
üìò *Libro enfocado en aprender C paso a paso, con ejercicios pr√°cticos y ejemplos claros sobre variables, operadores y estructuras de control.*

**4. ISO/IEC (2018)**  
**ISO/IEC 9899:2018 ‚Äî The C Programming Language Standard**  
üìò *Referencia oficial del lenguaje C. Ideal para consultar detalles sobre la sintaxis, comportamiento de funciones, tipos y operadores seg√∫n el est√°ndar.*

**5. Recursos online interactivos:**  
üåê [Learn-C.org](https://www.learn-c.org/) ‚Äì *Tutoriales pr√°cticos online, permite ejecutar c√≥digo directamente en el navegador.*  
üåê [Python Tutor ‚Äì Visualize C code](https://pythontutor.com/c.html#mode=edit) ‚Äì *Permite ver paso a paso c√≥mo se ejecuta el c√≥digo en C, ideal para entender variables, memoria y flujo de funciones.*  
üåê [Compilador on-line](https://www.onlinegdb.com/online_c_compiler) ‚Äì *Compila y ejecuta programas C directamente en el navegador, muy √∫til para practicar sin instalar nada.*


> *¬© 2025, Julia Clemente P√°rraga (<mailto:juliaclementep@gmail.com>). Licensed under [Creative Commons Attribution-NonCommercial-ShareAlike 4.0 International (CC BY-NC-SA 4.0)](https://creativecommons.org/licenses/by-nc-sa/4.0/).*
