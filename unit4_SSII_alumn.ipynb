# 2. Funciones y Módulos: Perspectiva Actual

En la programación moderna, las funciones y módulos no solo ayudan a organizar el código: son fundamentales para **claridad, reutilización y escalabilidad**, incluso en proyectos pequeños de C.  


### Diseño *Top-Down*  

- El **diseño Top-Down** consiste en dividir un problema grande en partes más pequeñas que se puedan resolver de forma separada.  
- Cada subproblema puede implementarse como:
  - **Función:** un bloque de código que realiza una tarea específica dentro del programa.
  - **Módulo:** un archivo o conjunto de funciones relacionadas que forman una unidad lógica, facilitando la organización y la reutilización.
- En ejercicios iniciales, cada trozo se implementa normalmente como una **función**; en programas más grandes, los trozos pueden agruparse en **módulos** que contengan varias funciones relacionadas.
- Esta estrategia permite centrarse primero en la **visión general del problema** y luego en la **implementación de cada pieza**, evitando perderse en detalles desde el inicio.


### Aplicación en programación

- Al aplicar Top-Down en programación:
  - Se divide el problema en **funciones y módulos** y se centra el trabajo en el desarrollo de cada uno.  
  - Beneficios de la modularización:
    - Facilita entender la **naturaleza del problema**.  
    - Permite centrar la atención en partes individuales antes de abordar el problema completo.  
    - Facilita la **prueba y depuración** de cada módulo por separado.  
    - Promueve la **reutilización** de código y la colaboración en equipo.  

- **En un programa en C:**
  - Solo debe haber una **función `main`**.  
  - La función `main` puede usar:  
    - **Librerías de funciones estándar** (`stdio.h`, `math.h`)  
    - **Otras funciones definidas en el programa**.  

- Una función en C consta de **dos partes principales**: la **declaración** y la **definición**.  
- La **declaración** indica al compilador que la función **existirá** y se podrá usar en el programa, incluso antes de conocer su implementación completa.  
- La **definición** es donde se escribe el código que realiza realmente la tarea de la función.  
- **Importante:** veremos la declaración y definición de funciones en detalle más adelante, cuando aprendamos a escribir funciones completas en C.   

### Buenas prácticas modernas

- **Funciones cortas y específicas:** cada función realiza **una sola tarea**.  
- **Módulos independientes:** cada módulo debería poder **compilarse y probarse por separado**.  
- **Documentación clara:** comentar la finalidad de cada función y los parámetros que recibe.  
- **Evitar efectos colaterales:** minimizar variables globales; mantener funciones auto-contenidas.  
- **Reutilización y testeo:** diseñar funciones que puedan ser usadas en otros programas.  
- **Prototipos y visibilidad:** declarar funciones antes de `main` mejora la legibilidad y evita errores de compilación.

#### 6.1.1 Enteros

Los **enteros** en C representan números **sin parte decimal**.  
El rango depende de:  

- El número de **bits** reservados para el tipo.  
- El modificador: `signed` (con signo) o `unsigned` (sin signo).  



💡 **Recuerda...**  
- Un tipo con `n` bits puede representar `2^n` valores distintos.  
- `unsigned` → usa todos los valores para no negativos (`0 … 2^n − 1`).  
- `signed` → divide el rango en negativos y positivos (`−2^(n−1) … 2^(n−1) − 1`).  
- Normalmente: `short ≤ int ≤ long ≤ long long` en número de bits.  
- Los tamaños concretos dependen del **compilador y la arquitectura** (Windows, Linux, 32 o 64 bits).  



**Tipos con signo (`signed`)**

| Tipo               | Bits típicos | Fórmula del rango             | Rango aproximado                  |
|--------------------|--------------|--------------------------------|-----------------------------------|
| `signed char`      | 8            | −2⁷ … 2⁷−1                   | −128 … 127                        |
| `signed short int` | 16           | −2¹⁵ … 2¹⁵−1                 | −32 768 … 32 767                  |
| `signed int`       | 32 (a veces 16) | −2³¹ … 2³¹−1                | −2 147 483 648 … 2 147 483 647    |
| `signed long int`  | 32 o 64      | −2^(n−1) … 2^(n−1)−1         | Windows (32 bits): como `int`<br>Linux 64 bits: −9 223 372 036 854 775 808 … 9 223 372 036 854 775 807 |


**Tipos sin signo (`unsigned`)**

| Tipo                  | Bits típicos | Fórmula del rango | Rango aproximado           |
|-----------------------|--------------|-------------------|----------------------------|
| `unsigned char`       | 8            | 0 … 2⁸−1         | 0 … 255                    |
| `unsigned short int`  | 16           | 0 … 2¹⁶−1        | 0 … 65 535                 |
| `unsigned int`        | 32           | 0 … 2³²−1        | 0 … 4 294 967 295          |
| `unsigned long int`   | 32 o 64      | 0 … 2ⁿ−1         | Windows (32 bits): 0 … 4 294 967 295<br>Linux 64 bits: 0 … 18 446 744 073 709 551 615 |


#### 6.1.2 Reales  

En C, los números **reales** se representan con los tipos `float` (32 bits) y `double` (64 bits), usando el estándar **IEEE 754**.

**Representación general (binaria):**  

Un número real tiene tres partes:  

- **Signo**: 1 bit (0 = positivo, 1 = negativo)  
- **Exponente**: ajusta la escala del número  
- **Mantisa (fracción)**: determina los dígitos significativos

<br> 

<div style="margin-top:10px; background-color:#d4edda; padding:8px; border-radius:5px; font-family: monospace;">
💡 <strong>Recuerda...</strong><br>
- Se representa en <strong>binario con notación científica</strong>: <code>(±1.mantisa) · 2^(exponente)</code>.<br>
- <code>float</code> → 32 bits, mantisa de 23 bits (~7 dígitos decimales confiables -sin pérdida de precisión-).<br>
- <code>double</code> → 64 bits, mantisa de 52 bits (~15 dígitos decimales confiables).<br>
- Los valores extremos se aproximan con <strong>potencias de 2</strong>:<br> 
  &nbsp;&nbsp;2²³ ≈ 8 388 608 → ~7 dígitos (`float`) <br>  
  &nbsp;&nbsp;2⁵² ≈ 4 503 599 627 370 496 → ~15 dígitos (`double`)<br>  
- Los números muy pequeños tienen el bit inicial en 0.<br>  
- Esto explica cuántos dígitos decimales podemos representar <strong>sin pérdida de precisión</strong>.
</div>

<br>

**Ejemplo: `float` (32 bits):**

Un `float` tiene: 1 bit de signo, 8 bits de exponente y 23 bits de mantisa **almacenados**.  
El **bit más significativo de la mantisa**, llamado **bit implícito**, **no se almacena**. Para los números cuyo valor permite usarlo (normalizados) se **asume que vale 1**.

| Valor aproximado       | Mantisa completa (bit implícito + almacenada) | Valor decimal aproximado       | Comentario |
|-----------------------|-----------------------------------------------|-------------------------------|------------|
| Muy grande positivo    | (1).11111111111111111111111b                  | +3.4028235 × 10<sup>38</sup> | Número positivo máximo |
| Pequeño positivo       | (1).00000000000000000000000b                  | +1.17549435 × 10<sup>-38</sup> | Número positivo pequeño |
| Muy cercano a 0        | (0).11111111111111111111111b                  | +1.40129846 × 10<sup>-45</sup> | Número muy pequeño |
| Cero                   | (0).00000000000000000000000b                  | 0                             | Cero absoluto |
| Cero negativo          | -(0).00000000000000000000000b                 | -0                            | Cero negativo (sign bit = 1) |
| Muy cercano a 0 negativo | -(0).11111111111111111111111b                | -1.40129846 × 10<sup>-45</sup> | Número muy pequeño negativo |
| Pequeño negativo       | -(1).00000000000000000000000b                 | -1.17549435 × 10<sup>-38</sup> | Número negativo pequeño |
| Muy grande negativo    | -(1).11111111111111111111111b                 | -3.4028235 × 10<sup>38</sup> | Número negativo máximo |

💡 **Recuerda:**  
- Cada bit en la mantisa **cuenta igual**, aunque los `1` puedan parecer en la tabla anterior visualmente más estrechos que los `0`.  
- La **mantisa “real”** tiene **24 bits**: 1 bit implícito + 23 bits almacenados.  
- El **bit implícito `(1)`** no se almacena, pero siempre se asume 1 para los números (normalizados).  
- El bit de signo determina si el número es positivo o negativo.

<br>  

> **Curiosidad, no evaluable**:  
> Valores especiales en IEEE 754:
>| Valor      | Significado breve |
>|------------|-----------------|
>| `+0`       | Cero positivo |
>| `-0`       | Cero negativo |
>| `+∞`       | Infinito positivo |
>| `-∞`       | Infinito negativo |
>| `NaN`      | Resultado de operación inválida (0/0, sqrt(-1)) |
>
> 
>
> - **Normalizado**: números “normales” que aprovechan todos los bits de la mantisa (bit implícito = 1).  
> - **Denormalizado**: números muy pequeños, cercanos a 0, con bit inicial = 0, permiten una transición suave hacia cero.

#### 6.3.1 Uso de `typedef`

El **`typedef`** permite crear un **alias** para un tipo existente, haciendo el código más **claro y compacto**.

💡 **Recuerda...**
- `typedef` **no crea un tipo nuevo**, solo un **alias**.
- Se puede usar con cualquier tipo, incluidos tipos definidos por el usuario como `struct` o `enum`.
- Facilita declarar variables de manera más **legible y uniforme**.

**Ejemplo 1: Alias de un tipo básico**  

<div style="text-align:center;">
  <div style="display:inline-block; font-family: monospace; background-color:#fff3cd; padding:8px 12px; border-radius:5px; font-size:16px;">
    <code>uchar</code> == <code>unsigned char</code>
  </div>
</div>

```c
#include <stdio.h>

typedef unsigned char uchar;

int main() {
    uchar a, b, c = 200;
    printf("Valor de c: %u\n", c);
    return 0;
}
```

```c
#include <stdio.h>

enum fruta { pera, limon, naranja };
typedef enum fruta Fruta;

int main() {
    Fruta sabor_del_zumo;

    printf("Introduce una fruta (0=pera, 1=limon, 2=naranja): ");
    scanf("%d", &sabor_del_zumo);

    printf("La fruta introducida es: %d\n", sabor_del_zumo);

    if (sabor_del_zumo == pera) printf("pera\n");
    else if (sabor_del_zumo == limon) printf("limon\n");
    else if (sabor_del_zumo == naranja) printf("naranja\n");

    return 0;
}



## 5. Agrupación de instrucciones  

En C, las instrucciones pueden agruparse para organizar mejor el código. Esto se hace mediante:

- **Sentencias simples**  
- **Bloques de sentencias**  
- **Funciones**  


### Sentencias simples

Una sentencia realiza una acción concreta:

```c
#include <stdio.h>

int main() {
    int a = 5, b = 3;
    int i;

    i = a + --b;  // sentencia simple
    printf("i = %d\n", i);

    return 0;
}
``` 

### Bloques de sentencias  

Un bloque agrupa varias sentencias entre `{ }`. Sirve para organizar código y limitar el alcance de las variables.  

```c
#include <stdio.h>

int main() {
    int total = 100;

    {
        int descuento = 20;          // variable local al bloque
        int subtotal = total - descuento;
        printf("Subtotal después del descuento: %d\n", subtotal);
    }

    // Aquí no se puede usar descuento ni subtotal
    printf("Total original: %d\n", total);

    return 0;
}
``` 

### Funciones   

Una función es un bloque de código reutilizable asociado a un nombre. Puede recibir parámetros y devolver un valor.  

<figure style="text-align: center;">
  <img src="figs/funcion_agrupacion.png" width="650"/>
  <figcaption>Fuente: Sistemas Informáticos, UAH (2025)</figcaption>
</figure>


<div class="alert alert-success" style="font-family: monospace;">
💡 <strong>RECUERDA...</strong><br>
- Una función puede aceptar <strong>cero o más parámetros</strong>.<br>
- Puede <strong>retornar un valor</strong> o ser <strong>void</strong> si no devuelve nada.<br>
- Permite <strong>reutilizar código</strong> y mantener el programa organizado.
</div>


## 6. Declaración y definición de una función  

<div style="background:#fff9e6; padding:15px; border-radius:8px; font-family: monospace; font-size: 1.25em;">

#include &lt;stdio.h&gt;<br><br>

<b>int Porcentaje (int tot, int pc);</b>   <b><span style="color:#006400;">// DECLARACIÓN o PROTOTIPO</span></b><br><br>

int main (void) {<br>
&nbsp;&nbsp;<b><span style="color:#006400;">// ...</span></b><br>
&nbsp;&nbsp;iva = <b>Porcentaje(</b>total, porcen<b>)</b>;   <b><span style="color:#006400;">// LLAMADA o EJECUCIÓN</span></b><br>
&nbsp;&nbsp;<b><span style="color:#006400;">// ...</span></b><br>
}<br><br>

<b>int Porcentaje (int tot, int pc)</b><br>
<b>{</b>   <b><span style="color:#006400;">// DEFINICIÓN o IMPLEMENTACIÓN</span></b><br>
&nbsp;&nbsp;<b><span style="color:#006400;">// ...</span></b><br>
<b>}</b><br>

</div>




## 7. Paso de parámetros a una función   


En C, las funciones pueden recibir **información desde el programa principal** u otras funciones mediante **parámetros**.  
Los parámetros permiten que una función **trabaje con datos distintos cada vez que se llama**, aumentando la **reutilización del código** y la **flexibilidad**.

Cuando se pasa un parámetro a una función, hay dos formas principales de hacerlo:

1. **Paso por valor:** la función recibe una copia del dato.  
2. **Paso por referencia:** la función recibe la dirección del dato, pudiendo modificar la variable original.

> 💡 Usar correctamente el tipo de paso de parámetros es fundamental para controlar **qué se modifica y qué no**, así como para optimizar memoria y eficiencia.


### 7.1 Paso de parámetros por valor  

<div style="background:#fff9e6; padding:15px; border-radius:8px; font-family: monospace; font-size: 1.25em;">

#include &lt;stdio.h&gt;<br><br>

<b>void intercambio(int x, int y);</b>   <b><span style="color:#006400;">// DECLARACIÓN o PROTOTIPO</span></b><br><br>

<b>int main(void)</b>   <b><span style="color:#006400;">// NUNCA usar void main()</span></b><br>
<b>{</b><br>
&nbsp;&nbsp;int a = 20;<br>
&nbsp;&nbsp;int b = 30;<br>
&nbsp;&nbsp;<b><span style="color:#006400;">// Mostramos valores y direcciones antes de la función</span></b><br>
&nbsp;&nbsp;printf("Main: a=%d (%p)\n", a, &a);<br>
&nbsp;&nbsp;printf("Main: b=%d (%p)\n", b, &b);<br>
<br>
&nbsp;&nbsp;<b><span style="background:#bbdefb; color:#000000;">// CLAVE: a → x, b → y (paso por valor)</span></b><br>
&nbsp;&nbsp;intercambio(a, b);<br>
<br>
&nbsp;&nbsp;<b><span style="color:#006400;">// Mostramos valores y direcciones después de la función</span></b><br>
&nbsp;&nbsp;printf("Main después: a=%d (%p)\n", a, &a);<br>
&nbsp;&nbsp;printf("Main después: b=%d (%p)\n", b, &b);<br>
&nbsp;&nbsp;return 0;<br>
<b>}</b><br><br>

<b>void intercambio(int x, int y)</b><br>
<b>{</b>   <b><span style="color:#006400;">// DEFINICIÓN o IMPLEMENTACIÓN</span></b><br>
<br>
&nbsp;&nbsp;<b><span style="background:#bbdefb; color:#000000;">// CLAVE: x y y reciben copias de los valores de a y b (a → x, b → y)</span></b><br>
<br>
&nbsp;&nbsp;<b><span style="color:#006400;">// Mostramos valores y direcciones al inicio</span></b><br>
&nbsp;&nbsp;printf("Antes swap: x=%d (%p)\n", x, &x);<br>
&nbsp;&nbsp;printf("Antes swap: y=%d (%p)\n", y, &y);<br>
<br>
&nbsp;&nbsp;<b><span style="color:#006400;">// Intercambiamos valores</span></b><br>
&nbsp;&nbsp;int temp;<br>
&nbsp;&nbsp;temp = x;<br>
&nbsp;&nbsp;x = y;<br>
&nbsp;&nbsp;y = temp;<br>
<br>
&nbsp;&nbsp;<b><span style="color:#006400;">// Mostramos valores y direcciones después del swap</span></b><br>
&nbsp;&nbsp;printf("Después swap: x=%d (%p)\n", x, &x);<br>
&nbsp;&nbsp;printf("Después swap: y=%d (%p)\n", y, &y);<br>
<b>}</b><br>

</div>

<div class="alert alert-info" style="font-family: monospace; font-size:1.1em;">
💡 <strong>Nota:</strong> ¿Funciona bien? Tecléalo y comprueba.
</div>



### **Visualización de memoria paso por valor:**   

<div style="font-family: monospace; font-size:1.25em;">

<h4>Antes de la llamada</h4>
<table style="border-collapse: collapse;">
<tr>
  <th style="border:1px solid black; padding:5px;">Variable</th>
  <th style="border:1px solid black; padding:5px;">Dirección</th>
  <th style="border:1px solid black; padding:5px;">Valor</th>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">a</td>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">0x7ffee1f4</td>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">20</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">b</td>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">0x7ffee1f0</td>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">30</td>
</tr>
</table>

<h4>Dentro de intercambio (inicio)</h4>
<table style="border-collapse: collapse;">
<tr>
  <th style="border:1px solid black; padding:5px;">Variable</th>
  <th style="border:1px solid black; padding:5px;">Dirección</th>
  <th style="border:1px solid black; padding:5px;">Valor</th>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#b9f6ca;">x</td>
  <td style="border:1px solid black; padding:5px; background:#b9f6ca;">0x7fffdf10</td>
  <td style="border:1px solid black; padding:5px; background:#b9f6ca;">20</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#69f0ae;">y</td>
  <td style="border:1px solid black; padding:5px; background:#69f0ae;">0x7fffdf0c</td>
  <td style="border:1px solid black; padding:5px; background:#69f0ae;">30</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#00c853;">temp</td>
  <td style="border:1px solid black; padding:5px; background:#00c853;">0x7fffdf08</td>
  <td style="border:1px solid black; padding:5px; background:#00c853;">-</td>
</tr>
</table>

<h4>Dentro de intercambio (después swap)</h4>
<table style="border-collapse: collapse;">
<tr>
  <th style="border:1px solid black; padding:5px;">Variable</th>
  <th style="border:1px solid black; padding:5px;">Dirección</th>
  <th style="border:1px solid black; padding:5px;">Valor</th>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#b9f6ca;">x</td>
  <td style="border:1px solid black; padding:5px; background:#b9f6ca;">0x7fffdf10</td>
  <td style="border:1px solid black; padding:5px; background:#b9f6ca;">30</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#69f0ae;">y</td>
  <td style="border:1px solid black; padding:5px; background:#69f0ae;">0x7fffdf0c</td>
  <td style="border:1px solid black; padding:5px; background:#69f0ae;">20</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#00c853;">temp</td>
  <td style="border:1px solid black; padding:5px; background:#00c853;">0x7fffdf08</td>
  <td style="border:1px solid black; padding:5px; background:#00c853;">20</td>
</tr>
</table>

<h4>Después de volver a main</h4>
<table style="border-collapse: collapse;">
<tr>
  <th style="border:1px solid black; padding:5px;">Variable</th>
  <th style="border:1px solid black; padding:5px;">Dirección</th>
  <th style="border:1px solid black; padding:5px;">Valor</th>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">a</td>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">0x7ffee1f4</td>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">20</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">b</td>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">0x7ffee1f0</td>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">30</td>
</tr>
</table>

</div>



<div class="alert alert-info" style="font-family: monospace; font-size:1.1em;">
💡 <strong>Nota:</strong> Se muestran solo las celdas implicadas en este ejemplo. Cada variable tiene su propia dirección, y puede imaginarse que están consecutivas en la <strong>pila</strong> (área de memoria donde se almacenan las variables locales de cada función y la dirección a la que debe volver la ejecución). Esto permite entender cómo funciona el <strong>paso por valor</strong>: las copias locales de la función no afectan a las variables originales en <strong>main</strong>.
</div>


### 7.2 Paso de parámetros por referencia   

<div style="background:#fff9e6; padding:15px; border-radius:8px; font-family: monospace; font-size: 1.25em;">

#include &lt;stdio.h&gt;<br><br>

<b>void intercambio(int *p, int *q);</b>   <b><span style="color:#006400;">// DECLARACIÓN o PROTOTIPO</span></b><br><br>

<b>int main(void)</b>   <b><span style="color:#006400;">// NUNCA usar void main()</span></b><br>
<b>{</b><br>
&nbsp;&nbsp;int a = 20;<br>
&nbsp;&nbsp;int b = 30;<br>
&nbsp;&nbsp;<b><span style="color:#006400;">// Mostramos valores y direcciones antes de la función</span></b><br>
&nbsp;&nbsp;printf("Main: a=%d (%p)\n", a, &a);<br>
&nbsp;&nbsp;printf("Main: b=%d (%p)\n", b, &b);<br>
<br>
&nbsp;&nbsp;<b><span style="background:#bbdefb; color:#000000;">// CLAVE: pasamos las direcciones de a y b (a → *p, b → *q)</span></b><br>
&nbsp;&nbsp;intercambio(&a, &b);<br>
<br>
&nbsp;&nbsp;<b><span style="color:#006400;">// Mostramos valores y direcciones después de la función</span></b><br>
&nbsp;&nbsp;printf("Main después: a=%d (%p)\n", a, &a);<br>
&nbsp;&nbsp;printf("Main después: b=%d (%p)\n", b, &b);<br>
&nbsp;&nbsp;return 0;<br>
<b>}</b><br><br>

<b>void intercambio(int *p, int *q)</b><br>
<b>{</b>   <b><span style="color:#006400;">// DEFINICIÓN o IMPLEMENTACIÓN</span></b><br>
<br>
&nbsp;&nbsp;<b><span style="background:#bbdefb; color:#000000;">// CLAVE: p apunta a a, q apunta a b (paso por referencia)</span></b><br>
<br>
&nbsp;&nbsp;<b><span style="color:#006400;">// Mostramos valores y direcciones al inicio</span></b><br>
&nbsp;&nbsp;printf("Antes swap: *p=%d (%p)\n", *p, p);<br>
&nbsp;&nbsp;printf("Antes swap: *q=%d (%p)\n", *q, q);<br>
&nbsp;&nbsp;<b><span style="color:#006400;">// Intercambiamos valores</span></b><br>
&nbsp;&nbsp;int temp;<br>
&nbsp;&nbsp;temp = *p;<br>
&nbsp;&nbsp;*p = *q;<br>
&nbsp;&nbsp;*q = temp;<br>
&nbsp;&nbsp;<b><span style="color:#006400;">// Mostramos valores y direcciones después del swap</span></b><br>
&nbsp;&nbsp;printf("Después swap: *p=%d (%p)\n", *p, p);<br>
&nbsp;&nbsp;printf("Después swap: *q=%d (%p)\n", *q, q);<br>
<b>}</b><br>

</div>


### **Visualización de memoria paso por referencia:**   

<div style="font-family: monospace; font-size:1.25em;">

<h4>Antes de la llamada</h4>
<table style="border-collapse: collapse;">
<tr>
  <th style="border:1px solid black; padding:5px;">Variable</th>
  <th style="border:1px solid black; padding:5px;">Dirección</th>
  <th style="border:1px solid black; padding:5px;">Contenido / Valor</th>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">a</td>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">0x7ffee1f4</td>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">20</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">b</td>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">0x7ffee1f0</td>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">30</td>
</tr>
</table>

<h4>Dentro de intercambio (inicio)</h4>
<table style="border-collapse: collapse;">
<tr>
  <th style="border:1px solid black; padding:5px;">Variable</th>
  <th style="border:1px solid black; padding:5px;">Dirección</th>
  <th style="border:1px solid black; padding:5px;">Contenido / Valor</th>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">a</td>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">0x7ffee1f4</td>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">20</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">b</td>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">0x7ffee1f0</td>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">30</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#b9f6ca;">p</td>
  <td style="border:1px solid black; padding:5px; background:#b9f6ca;">0x7fffdf10</td>
  <td style="border:1px solid black; padding:5px; background:#b9f6ca;">0x7ffee1f4</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#69f0ae;">q</td>
  <td style="border:1px solid black; padding:5px; background:#69f0ae;">0x7fffdf0c</td>
  <td style="border:1px solid black; padding:5px; background:#69f0ae;">0x7ffee1f0</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#00c853;">temp</td>
  <td style="border:1px solid black; padding:5px; background:#00c853;">0x7fffdf08</td>
  <td style="border:1px solid black; padding:5px; background:#00c853;">-</td>
</tr>
</table>

<h4>Dentro de intercambio (después swap)</h4>
<table style="border-collapse: collapse;">
<tr>
  <th style="border:1px solid black; padding:5px;">Variable</th>
  <th style="border:1px solid black; padding:5px;">Dirección</th>
  <th style="border:1px solid black; padding:5px;">Contenido / Valor</th>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">a</td>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">0x7ffee1f4</td>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">30</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">b</td>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">0x7ffee1f0</td>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">20</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#b9f6ca;">p</td>
  <td style="border:1px solid black; padding:5px; background:#b9f6ca;">0x7fffdf10</td>
  <td style="border:1px solid black; padding:5px; background:#b9f6ca;">0x7ffee1f4</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#69f0ae;">q</td>
  <td style="border:1px solid black; padding:5px; background:#69f0ae;">0x7fffdf0c</td>
  <td style="border:1px solid black; padding:5px; background:#69f0ae;">0x7ffee1f0</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#00c853;">temp</td>
  <td style="border:1px solid black; padding:5px; background:#00c853;">0x7fffdf08</td>
  <td style="border:1px solid black; padding:5px; background:#00c853;">30</td>
</tr>
</table>

<h4>Después de volver a main</h4>
<table style="border-collapse: collapse;">
<tr>
  <th style="border:1px solid black; padding:5px;">Variable</th>
  <th style="border:1px solid black; padding:5px;">Dirección</th>
  <th style="border:1px solid black; padding:5px;">Contenido / Valor</th>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">a</td>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">0x7ffee1f4</td>
  <td style="border:1px solid black; padding:5px; background:#fff59d;">30</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">b</td>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">0x7ffee1f0</td>
  <td style="border:1px solid black; padding:5px; background:#ffcc80;">20</td>
</tr>
</table>

</div>

<div class="alert alert-info" style="font-family: monospace; font-size:1.1em;">
💡 <strong>Nota:</strong> Se muestran solo las celdas implicadas en este ejemplo. Cada variable tiene su propia dirección, y puede imaginarse que están consecutivas en la <strong>pila</strong>. Los punteros <strong>p</strong> y <strong>q</strong> contienen las direcciones de <strong>a</strong> y <strong>b</strong>, y al modificarlos mediante <strong>*p</strong> y <strong>*q</strong>, los valores originales cambian en <strong>main</strong>.
</div>


## 8. Programa formado por varios archivos  

Cuando un programa crece, conviene **separar las funciones en varios archivos** para organizar mejor el código y facilitar su mantenimiento.

### Archivo 1: `modulo01.c` (función `main`)

```c
// modulo01.c: función main
#include <stdio.h>

// Declaración de la función Media
double Media(double, double);

int main(void)  // nunca void main()
{
    double x = 2.34;
    double y = 1.73;
    double m;

    // Llamada a la función Media
    m = Media(x, y);

    printf("Media == %f\n", m);

    return 0;
}
```

## 10. Visibilidad de variables

En C, la **visibilidad** de una variable depende de dónde se declare:

<div class="alert alert-primary" style="font-family: monospace; font-size:1.1em;">
🟦 <strong>Variables globales</strong><br>
- Se declaran fuera de cualquier función.<br>
- Son visibles desde su definición en adelante, en todas las funciones del archivo.<br>
- Útiles para información compartida.
</div>

<div class="alert alert-success" style="font-family: monospace; font-size:1.1em;">
🟩 <strong>Variables locales</strong><br>
- Se declaran dentro de un bloque (por ejemplo, dentro de `main()` o un `if`).<br>
- Solo son visibles dentro de ese bloque y sus sub-bloques.
</div>

<div class="alert alert-warning" style="font-family: monospace; font-size:1.1em;">
🟨 <strong>Variables con nombres iguales</strong><br>
- Pueden existir en bloques diferentes.<br>
- Si un bloque es sub-bloque de otro, la variable del sub-bloque <strong>eclipsa</strong> a la del bloque superior.<br>
- El tipo puede ser distinto, no importa.
</div>


### Ejemplo sencillo:

```c
#include <stdio.h>

// Variable global
int global = 10;

int main(void)
{
    int local = 5; // Variable local en main

    printf("Global: %d, Local: %d\n", global, local);

    {   // Inicio de un sub-bloque
        int local = 20; // Eclipsa a la variable 'local' de main
        int sub = 7;
        printf("Dentro sub-bloque -> Local eclipsada: %d, Sub: %d\n", local, sub);
    }

    // Aquí 'sub' ya no existe, pero 'local' de main sigue visible
    printf("Después del sub-bloque -> Local: %d\n", local);

    return 0;
}


## 11. Calificación de variables

En C, los calificadores modifican el **comportamiento de las variables**, afectando su **duración, visibilidad y forma de acceso**.  

### Tabla de calificación

<table style="border-collapse: collapse; font-family: monospace; font-size:1.1em;">
<tr>
  <th style="border:1px solid black; padding:5px;">Calificador</th>
  <th style="border:1px solid black; padding:5px;">Local</th>
  <th style="border:1px solid black; padding:5px;">Global</th>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px;">auto (por defecto)</td>
  <td style="border:1px solid black; padding:5px;">Se crea, inicializa y destruye cada vez que se ejecuta el bloque.</td>
  <td style="border:1px solid black; padding:5px;">No aplicable</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px;">static</td>
  <td style="border:1px solid black; padding:5px;">Permanece entre llamadas; inicializa solo la primera vez.</td>
  <td style="border:1px solid black; padding:5px;">Accesible solo desde el archivo donde se define.</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px;">extern</td>
  <td style="border:1px solid black; padding:5px;">No aplicable</td>
  <td style="border:1px solid black; padding:5px;">Indica que la variable global está definida en otro archivo o en otro lugar del mismo archivo.</td>
</tr>
<tr>
  <td style="border:1px solid black; padding:5px;">volatile</td>
  <td style="border:1px solid black; padding:5px;">No almacenar en registros ni en caché del compilador; cada lectura/escritura va directamente a memoria principal.</td>
  <td style="border:1px solid black; padding:5px;">Igual que local: no almacenar en registros; siempre leer/escribir en memoria principal.</td>
</tr>
</table>

### Ejemplo práctico

```c
#include <stdio.h>

int global = 5;          // variable global normal
static int archivo = 10; // variable global estática

int main(void)
{
    auto int local = 3;      // auto, por defecto
    static int contador = 0; // local estática, persiste entre llamadas
    volatile int sensor = 7; // local volatile

    contador++;
    printf("Global: %d, Local: %d, Contador: %d, Sensor: %d\n",
           global, local, contador, sensor);

    return 0;
}


## 12. 📚 Lecturas Recomendadas

Estas referencias complementan el contenido del notebook y ayudan a los alumnos a profundizar o contextualizar lo visto:

**1. Deitel, P. J., & Deitel, H. M. (2022)**  
**C: How to Program (10th ed.)**, Pearson.  
📘 *Referencia principal para ejemplos de programación y comprensión de cómo se representa la información en un ordenador. Útil para ver la relación entre números y caracteres.*

**2. Ceballos Sierra, F. J. (2019)**  
**C/C++. Curso de programación (5.ª ed.)**, RA-MA Editorial.  
📘 *Muy útil para repasar la sintaxis y estructuras de C, con ejemplos prácticos que complementan los conceptos vistos en el notebook.*

**3. Packt Publishing (2019)**  
**Learn C Programming**, Packt Publishing.  
📘 *Libro enfocado en aprender C paso a paso, con ejercicios prácticos y ejemplos claros sobre variables, operadores y estructuras de control.*

**4. ISO/IEC (2018)**  
**ISO/IEC 9899:2018 — The C Programming Language Standard**  
📘 *Referencia oficial del lenguaje C. Ideal para consultar detalles sobre la sintaxis, comportamiento de funciones, tipos y operadores según el estándar.*

**5. Recursos online interactivos:**  
🌐 [Learn-C.org](https://www.learn-c.org/) – *Tutoriales prácticos online, permite ejecutar código directamente en el navegador.*  
🌐 [Python Tutor – Visualize C code](https://pythontutor.com/c.html#mode=edit) – *Permite ver paso a paso cómo se ejecuta el código en C, ideal para entender variables, memoria y flujo de funciones.*  
🌐 [Compilador on-line](https://www.onlinegdb.com/online_c_compiler) – *Compila y ejecuta programas C directamente en el navegador, muy útil para practicar sin instalar nada.*


> *© 2025, Julia Clemente Párraga (<mailto:juliaclementep@gmail.com>). Licensed under [Creative Commons Attribution-NonCommercial-ShareAlike 4.0 International (CC BY-NC-SA 4.0)](https://creativecommons.org/licenses/by-nc-sa/4.0/).*
