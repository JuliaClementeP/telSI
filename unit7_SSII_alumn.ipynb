## 1. Introducci√≥n

# 2. Estructuras: declaraci√≥n y definici√≥n  

Antes de poder utilizar una estructura en un programa en C, es necesario **declararla** y posteriormente **definir** variables basadas en ella.  

La **declaraci√≥n de una estructura** indica **qu√© campos la componen** y **de qu√© tipo** es cada uno, es decir, describe el "molde" o "plantilla" que define c√≥mo ser√° la estructura.  

Por su parte, la **definici√≥n de una estructura** consiste en **crear variables** que utilizan ese molde, permitiendo almacenar valores concretos en cada uno de sus campos.  

Estas dos etapas ‚Äîdeclarar y definir‚Äî son fundamentales para poder trabajar correctamente con estructuras, ya que permiten **organizar y manipular datos heterog√©neos** bajo un mismo nombre de tipo.

Antes de poder utilizar una estructura en un programa, es necesario **declarar su tipo**, es decir, indicar **qu√© campos la forman** y **de qu√© tipo de dato** es cada uno.  

La **declaraci√≥n de una estructura** define un nuevo tipo compuesto que agrupa diferentes tipos de datos bajo un mismo nombre. Esta declaraci√≥n act√∫a como un *molde* o *plantilla* a partir de la cual, m√°s adelante, se podr√°n crear variables de ese tipo.

### 2.2 Definici√≥n de estructuras

### 2.3 D√≥nde ubicar las declaraciones de estructuras

#### 2) Declaraci√≥n en un archivo de cabecera (`.h`)

## 3. Lectura de los datos de una estructura

En esta secci√≥n veremos c√≥mo **leer los datos de una estructura** directamente desde el `main` usando `scanf` y `fgets`.  
Tomando como base nuestro ejemplo de la estructura `ficha`, podemos hacerlo de la siguiente manera:

```c

    struct ficha estudiante;
    int ret;

    // Lectura de datos b√°sicos
    printf("Introduzca los datos del estudiante:\n");

    printf("Nombre: ");
    fgets(estudiante.nom, sizeof(estudiante.nom), stdin);

    printf("Direcci√≥n: ");
    fgets(estudiante.dir, sizeof(estudiante.dir), stdin);

    printf("Tel√©fono: ");
    scanf("%ld", &estudiante.telefono);

    // Limpieza del buffer antes de leer la fecha
    while (getchar() != '\n');

    // Lectura de la fecha de nacimiento
    printf("Fecha de nacimiento (dd/mm/aaaa): ");
    do {
        ret = scanf("%d/%d/%d", 
                     &estudiante.nac.dia, 
                     &estudiante.nac.mes, 
                     &estudiante.nac.anyo);
        while (getchar() != '\n');  // limpiar buffer de entrada
    } while (ret != 3);

    // Mostrar los datos ingresados
    printf("\n--- Datos del estudiante ---\n");
    printf("Nombre: %s", estudiante.nom);
    printf("Direcci√≥n: %s", estudiante.dir);
    printf("Tel√©fono: %ld\n", estudiante.telefono);
    printf("Fecha de nacimiento: %02d/%02d/%04d\n",
           estudiante.nac.dia, estudiante.nac.mes, estudiante.nac.anyo);

    return 0;
}
```

> ‚ö†Ô∏è No te olvides de lo ya visto: Siempre se debe limpiar el buffer con `while(getchar() != '\n');` (m√≠nimo) antes de pasar de `scanf()` a `fgets()` para evitar que queden caracteres de nueva l√≠nea y entre determinados `scanf()`.

## 5. Copia de estructuras  

En C, las estructuras pueden copiarse directamente entre s√≠ usando el operador de asignaci√≥n (`=`), siempre que ambas estructuras sean del mismo tipo. Esto permite duplicar el contenido completo (campo a campo) de una estructura a otra de manera sencilla.  

<div class="alert alert-success" style="font-family: monospace;">
üí° <strong>Concepto clave</strong><br>
Cuando hacemos una asignaci√≥n entre estructuras del mismo tipo, <strong>C copia todos los valores de sus miembros</strong>.<br>
Es una <strong>copia completa y por valor</strong>, no una referencia: modificar una no afecta a la otra.
</div>


### Ejemplo de c√≥digo  

```c
#include "ficha.h"

    struct ficha persona1, persona2;

    // Se leen los datos de la primera persona
    persona1 = leerFichav1();

    // Se copia persona1 en persona2
    persona2 = persona1;

    // Mostramos los datos de la copia
    visualizarFicha(persona2);

    return 0;
}
```

> ‚ö†Ô∏è Importante:  
> Esta copia solo es v√°lida si las estructuras son del mismo tipo.
> Si una estructura contiene punteros, la copia ser√° superficial: solo se copiar√°n las direcciones, no los datos apuntados. En esos casos, puede ser necesario realizar una copia profunda (`deep copy`) con c√≥digo adicional.

## Ejemplo:

## 6. Acceso a los campos de una estructura  

En C, una vez que tenemos una estructura declarada, podemos acceder a sus campos de dos maneras principales:

<div class="alert alert-success" style="font-family: monospace;">

- Para acceder a sus campos:<br>
    * Si tenemos la variable, usamos <code>.</code> (punto).<br>   
    * Si tenemos un puntero a la variable, usamos <code>-></code> (flecha).<br>

### Ejemplo:

    // Declaraci√≥n de una estructura y un puntero

````python
/*------------------------------------------------------------
 * Fichero: main.c
 * Autor: [Aqu√≠ tu nombre]
 * Fecha: [Fecha actual]
 *
 * Descripci√≥n:
 *   Programa principal que permite al usuario interactuar con
 *   un conjunto de p√≥lizas de seguro mediante un men√∫.
 *   Utiliza las funciones del m√≥dulo poliza.c para realizar
 *   las operaciones de lectura, b√∫squeda, modificaci√≥n,
 *   c√°lculo del valor y ordenaci√≥n.
 *
 *------------------------------------------------------------*/

#include <stdio.h>
#include <string.h>
#include "poliza.h"

#define MAX 50

int main(void) {
    tDatosPoliza polizas[MAX];
    int n = 0;
    int opcion, anyoActual = 2025;
    char numBuscar[50];

    do {
        printf("\n===== MENU PRINCIPAL =====\n");
        printf("1. Introducir nueva p√≥liza\n");
        printf("2. Mostrar todas las p√≥lizas\n");
        printf("3. Calcular valor de p√≥lizas\n");
        printf("4. Buscar p√≥liza por n√∫mero\n");
        printf("5. Modificar p√≥liza\n");
        printf("6. Ordenar por matr√≠cula\n");
        printf("7. Salir\n");
        printf("Seleccione una opci√≥n: ");
        scanf("%d", &opcion);
        limpiarBuffer();

        switch (opcion) {
            case 1:
                if (n < MAX) {
                    leerPolizaV2(&polizas[n]);
                    n++;
                } else {
                    printf("No se pueden introducir m√°s p√≥lizas.\n");
                }
                break;

            case 2:
                for (int i = 0; i < n; i++)
                    mostrarPoliza(&polizas[i]);
                break;

            case 3:
                for (int i = 0; i < n; i++)
                    calcularValorPoliza(&polizas[i], anyoActual);
                printf("Valores de p√≥lizas actualizados.\n");
                break;

            case 4:
                printf("Introduzca el n√∫mero de p√≥liza a buscar: ");
                fgets(numBuscar, sizeof(numBuscar), stdin);
                numBuscar[strcspn(numBuscar, "\n")] = '\0';
                {
                    int pos = buscarPoliza(polizas, n, numBuscar);
                    if (pos >= 0)
                        mostrarPoliza(&polizas[pos]);
                    else
                        printf("P√≥liza no encontrada.\n");
                }
                break;

            case 5:
                printf("N√∫mero de p√≥liza a modificar: ");
                fgets(numBuscar, sizeof(numBuscar), stdin);
                numBuscar[strcspn(numBuscar, "\n")] = '\0';
                {
                    int pos = buscarPoliza(polizas, n, numBuscar);
                    if (pos >= 0)
                        modificarPoliza(&polizas[pos]);
                    else
                        printf("P√≥liza no encontrada.\n");
                }
                break;

            case 6:
                ordenarPorMatricula(polizas, n);
                break;

            case 7:
                printf("Saliendo del programa...\n");
                break;

            default:
                printf("Opci√≥n no v√°lida.\n");
        }

    } while (opcion != 7);

    return 0;
}
````
