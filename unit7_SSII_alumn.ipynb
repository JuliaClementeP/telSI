## 1. Introducción   

En los programas en C, con frecuencia es necesario manejar conjuntos de datos que pertenecen a un mismo objeto o entidad, pero que no son del mismo tipo. Por ejemplo, para representar la información de un estudiante podríamos necesitar almacenar su **nombre** (cadena de caracteres), **edad** (entero) y **promedio** (real).  

Una **estructura** permite agrupar todos esos datos relacionados bajo un mismo identificador, es decir:    

- Una **estructura** es un elemento que almacena, bajo un identificador común, **varios tipos de datos** que pueden ser de distinta naturaleza (enteros, reales, caracteres, etc.).  
- Para poder usar una estructura, primero hay que **declararla**, indicando qué **campos** (o miembros) la forman y **de qué tipo** es cada uno de ellos.  

Las estructuras son fundamentales para representar información más compleja y organizar los datos de forma clara y coherente. A lo largo de este tema aprenderás a:  

- **Declarar y definir estructuras**,  
- **Leer y modificar sus datos** desde funciones,  
- **Acceder a sus campos** correctamente,  
- **Crear sinónimos** con `typedef`,  
- **Trabajar con arrays de estructuras**,  
- Y finalmente, **aplicar todos estos conceptos** en un problema práctico paso a paso.  

El dominio de las estructuras constituye un paso clave en el aprendizaje del lenguaje C, ya que sienta las bases para la **programación modular** y la **organización eficiente de datos** en programas más complejos.


# 2. Estructuras: declaración y definición  

Antes de poder utilizar una estructura en un programa en C, es necesario **declararla** y posteriormente **definir** variables basadas en ella.  

La **declaración de una estructura** indica **qué campos la componen** y **de qué tipo** es cada uno, es decir, describe el "molde" o "plantilla" que define cómo será la estructura.  

Por su parte, la **definición de una estructura** consiste en **crear variables** que utilizan ese molde, permitiendo almacenar valores concretos en cada uno de sus campos.  

Estas dos etapas —declarar y definir— son fundamentales para poder trabajar correctamente con estructuras, ya que permiten **organizar y manipular datos heterogéneos** bajo un mismo nombre de tipo.  

### 2.1 Declaración de estructuras   

Antes de poder utilizar una estructura en un programa, es necesario **declarar su tipo**, es decir, indicar **qué campos la forman** y **de qué tipo de dato** es cada uno.  

La **declaración de una estructura** define un nuevo tipo compuesto que agrupa diferentes tipos de datos bajo un mismo nombre. Esta declaración actúa como un *molde* o *plantilla* a partir de la cual, más adelante, se podrán crear variables de ese tipo.  

### Forma genérica  

A continuación se indica la forma genérica de declarar una estructura:  

```c
struct identificador {
    tipo_dato identificador_miembro_1;
    tipo_dato identificador_miembro_2;
    /* … */
    tipo_dato identificador_miembro_m;
};
```

> Observa los elementos de la declaración de una estructura.
> ⚠️ En esta etapa solo hemos declarado el tipo, **no se crean las variables**.

<div class="alert alert-success" style="font-family: monospace;">
💡 <strong>Recuerda...</strong><br>
- Una <strong>estructura</strong> agrupa diferentes tipos de datos bajo un mismo identificador.<br>
- Sirve como <strong>molde o plantilla</strong> para crear variables que contendrán valores concretos.<br>
- Todavía <strong>no se han creado variables</strong>: eso se hará en la siguiente sección (<strong>2.2</strong>).<br>  
    
<br>

<strong>¡Buenas prácticas relativas a la declaración de estructuras!</strong><br>

- Coloca las declaraciones de estructuras al inicio del archivo fuente (<code>.c</code>) o en un archivo de cabecera (<code>.h</code>), especialmente si serán utilizadas por varias funciones o módulos.

- Utiliza nombres descriptivos para los campos (por ejemplo, <code>edad</code>, <code>promedio</code>, <code>nombre</code> en lugar de <code>a</code>, <code>b</code>, <code>c</code>).

- Evita usar palabras reservadas del lenguaje como nombres de campos o estructuras.
</div>


### 2.2 Definición de estructuras  

Una vez que hemos **declarado** un tipo de estructura (es decir, creado su "molde"), podemos **definir variables** de ese tipo.  
Definir una variable significa **crear un espacio en memoria** que contendrá valores concretos para cada campo de la estructura.  

La sintaxis básica para definir variables de un tipo `struct` es:

```c
struct identificador variable_1, variable_2, ...;
```

**Ejemplo completo con estructuras anidadas:**  

Primero declaramos los tipos de estructuras:  

```c
// Declaración de tipo fecha
struct fecha {
    int dia;
    int mes;
    int anyo;
};

// Declaración de tipo ficha, que incluye un campo de tipo struct fecha
struct ficha {
    char nom[40];
    char dir[100];
    long telefono;
    struct fecha nac; // fecha de nacimiento
};
```
> ⚠️ Lo vas a ver muy a menudo en el tema, pero recuerda: la declaración del tipo siempre debe estar fuera de main o antes de cualquier función que la utilice, para que el compilador conozca la estructura.


**Definición e inicialización de variables:**  

```c
struct fecha hoy = {4, 10, 2019}; //variable hoy de tipo struct fecha
```
**Valor de cada campo:**  

```c
hoy.dia = 4

hoy.mes = 10

hoy.anyo = 2019
```

> De momento estamos viendo aspectos no integrados en un programa en C, sólo su sintaxis y semántica aisladas.

```c
struct ficha estudiante = {  //variable de tipo struct ficha
    "Lucas",
    "Calle del Cerezo 4",
    910002233,
    {1, 3, 2000}  // fecha de nacimiento: dia, mes, anyo
};
```

```c
estudiante.nom = "Lucas"

estudiante.dir = "Calle del Cerezo 4"

estudiante.telefono = 910002233

estudiante.nac.dia = 1

estudiante.nac.mes = 3

estudiante.nac.anyo = 2000
```

> ⚠️ Nota: Para estructuras anidadas, como la anterior, los campos internos se inicializan en el mismo orden en que fueron declarados.

<div class="alert alert-success" style="font-family: monospace;">
💡 <strong>Recuerda siempre seguir "Buenas prácticas" al definir estructuras...</strong><br>
- Utiliza nombres <strong>claros y descriptivos</strong> para variables y campos.<br>
- Inicializa las variables en el momento de la definición siempre que sea posible.<br>
- Para <strong>estructuras anidadas</strong>, respeta el orden de los campos al inicializar.<br>
- Cuando incluyas <strong>arrays de caracteres</strong> (<code>char[]</code>), recuerda que deben ser cadenas terminadas en <code>'\0'</code>.<br>
- Mantén el código limpio y bien indentado para facilitar la lectura y evitar errores.  
</div>


### 2.3 Dónde ubicar las declaraciones de estructuras

En C, las estructuras **deben declararse antes de ser utilizadas** en cualquier función o definición de variable.  
Existen dos formas habituales de organizarlas:

1. **Directamente en el archivo fuente (`.c`)**, antes de `main()`.  
2. **En un archivo de cabecera (`.h`)**, que luego se incluye con `#include` en el archivo principal.  

#### 1) Declaración directamente en el archivo fuente

```c
#include <stdio.h>

// Declaración de estructuras
struct fecha {
    int dia, mes, anyo;
};

struct ficha {
    char nom[40];
    char dir[40];
    long telefono;
    struct fecha nac;
};

int main() {
    // Definición e inicialización de variables
    struct fecha hoy = {4, 10, 2019};
    struct ficha estudiante = {
        "Lucas",
        "Calle del Cerezo 4",
        910002233,
        {1, 3, 2000}
    };

    // Resto del programa
    return 0;
}
```

#### 2) Declaración en un archivo de cabecera (`.h`)  

Para proyectos más grandes, es recomendable separar **las declaraciones de estructuras** en un archivo de cabecera para poder reutilizarlas en varios archivos fuente.  

**`ficha.h`**   

<div class="alert alert-info" style="font-family: monospace;">

```c
#ifndef _FICHA_H
#define _FICHA_H

struct fecha {
    int dia, mes, anyo;  
};

struct ficha {  
    char nom[40];  
    char dir[40];  
    long telefono;  
    struct fecha nac;  
};

#endif
```
</div>

**`main.c`**

```c
#include <stdio.h>
#include "ficha.h"

int main() {
    struct ficha estudiante;

    // Resto del código
    return 0;
}
```

> - Atento que te cuento brevemente para qué sirven: `#ifndef _FICHA_H ... #define _FICHA_H ... #endif`.  
> - ¿Dónde pueden ser utilizadas las estructuras declaradas en `ficha.h`?  
> - Mantener las declaraciones de estructuras en archivos `.h` permite que el código sea **modular, limpio y reutilizable**.  
> - Siempre coloca la **declaración de estructuras** **antes de definir variables** que las utilicen, ya sea en el mismo archivo `.c` o mediante un *header* `.h`.

## 3. Lectura de los datos de una estructura

En esta sección veremos cómo **leer los datos de una estructura** directamente desde el `main` usando `scanf` y `fgets`.  
Tomando como base nuestro ejemplo de la estructura `ficha`, podemos hacerlo de la siguiente manera:

```c
#include <stdio.h>
#include "ficha.h"

int main() {
    struct ficha estudiante;
    int ret;

    // Lectura de datos básicos
    printf("Introduzca los datos del estudiante:\n");

    printf("Nombre: ");
    fgets(estudiante.nom, sizeof(estudiante.nom), stdin);

    printf("Dirección: ");
    fgets(estudiante.dir, sizeof(estudiante.dir), stdin);

    printf("Teléfono: ");
    scanf("%ld", &estudiante.telefono);

    // Limpieza del buffer antes de leer la fecha
    while (getchar() != '\n');

    // Lectura de la fecha de nacimiento
    printf("Fecha de nacimiento (dd/mm/aaaa): ");
    do {
        ret = scanf("%d/%d/%d", 
                     &estudiante.nac.dia, 
                     &estudiante.nac.mes, 
                     &estudiante.nac.anyo);
        while (getchar() != '\n');  // limpiar buffer de entrada
    } while (ret != 3);

    // Mostrar los datos ingresados
    printf("\n--- Datos del estudiante ---\n");
    printf("Nombre: %s", estudiante.nom);
    printf("Dirección: %s", estudiante.dir);
    printf("Teléfono: %ld\n", estudiante.telefono);
    printf("Fecha de nacimiento: %02d/%02d/%04d\n",
           estudiante.nac.dia, estudiante.nac.mes, estudiante.nac.anyo);

    return 0;
}
```

> ⚠️ No te olvides de lo ya visto: Siempre se debe limpiar el buffer con `while(getchar() != '\n');` (mínimo) antes de pasar de `scanf()` a `fgets()` para evitar que queden caracteres de nueva línea.

## 3.1 Lectura de los datos de una estructura en una función

En lugar de leer los datos directamente en `main`, podemos **usar una función** para encapsular la lógica de entrada. Esto mejora la organización del código y facilita la reutilización.


### Versión 1: función que retorna la estructura  

```c
#include <stdio.h>
#include "ficha.h"

struct ficha leerFichav1() {
    struct ficha estudiante;
    int ret;

    printf("Introduzca los datos del estudiante:\n");

    printf("Nombre: ");
    fgets(estudiante.nom, sizeof(estudiante.nom), stdin);

    printf("Dirección: ");
    fgets(estudiante.dir, sizeof(estudiante.dir), stdin);

    printf("Teléfono: ");
    scanf("%ld", &estudiante.telefono);

    // Limpiar buffer antes de leer la fecha
    while (getchar() != '\n');

    printf("Fecha de nacimiento (dd/mm/aaaa): ");
    do {
        ret = scanf("%d/%d/%d",
                     &estudiante.nac.dia,
                     &estudiante.nac.mes,
                     &estudiante.nac.anyo);
        while (getchar() != '\n'); // limpiar buffer
    } while (ret != 3);

    return estudiante;
}

int main() {
    struct ficha estudiante;

    estudiante = leerFichav1();

    // Mostrar los datos
    printf("\n--- Datos del estudiante ---\n");
    printf("Nombre: %s", estudiante.nom);
    printf("Dirección: %s", estudiante.dir);
    printf("Teléfono: %ld\n", estudiante.telefono);
    printf("Fecha de nacimiento: %02d/%02d/%04d\n",
           estudiante.nac.dia, estudiante.nac.mes, estudiante.nac.anyo);

    return 0;
}
```

### Versión 2: función que recibe un puntero a la estructura  

```c
#include <stdio.h>
#include "ficha.h"

void leerFichav2(struct ficha *estudiante) {
    int ret;

    printf("Introduzca los datos del estudiante:\n");

    printf("Nombre: ");
    fgets(estudiante->nom, sizeof(estudiante->nom), stdin);

    printf("Dirección: ");
    fgets(estudiante->dir, sizeof(estudiante->dir), stdin);

    printf("Teléfono: ");
    scanf("%ld", &estudiante->telefono);

    // Limpiar buffer antes de leer la fecha
    while (getchar() != '\n');

    printf("Fecha de nacimiento (dd/mm/aaaa): ");
    do {
        ret = scanf("%d/%d/%d",
                     &estudiante->nac.dia,
                     &estudiante->nac.mes,
                     &estudiante->nac.anyo);
        while (getchar() != '\n'); // limpiar buffer
    } while (ret != 3);
}

int main() {
    struct ficha estudiante;

    leerFichav2(&estudiante);

    // Mostrar los datos
    printf("\n--- Datos del estudiante ---\n");
    printf("Nombre: %s", estudiante.nom);
    printf("Dirección: %s", estudiante.dir);
    printf("Teléfono: %ld\n", estudiante.telefono);
    printf("Fecha de nacimiento: %02d/%02d/%04d\n",
           estudiante.nac.dia, estudiante.nac.mes, estudiante.nac.anyo);

    return 0;
}
```
> 💡 **Nota importante:**  
> Usamos `->` para acceder a los campos de una estructura **a través de un puntero**, mientras que con variables normales usamos `.`.

### 3.2 Invocación a la función

En esta sección veremos cómo **invocar las funciones que leen datos de una estructura** desde `main`.  
Para que quede más claro, mostraremos **dos programas separados**, uno para cada versión de la función:

- **Versión 1:** la función **retorna la estructura**, y asignamos la copia a una variable.  
- **Versión 2:** la función **recibe un puntero** a la estructura y modifica directamente la variable definida en `main`.

#### Versión 1: función que retorna la estructura

```c
#include <stdio.h>
#include "ficha.h"

// Declaración
struct ficha leerFichav1();

int main() {
    struct ficha persona1;

    persona1 = leerFichav1();

    printf("\n--- Persona 1 ---\n");
    printf("Nombre: %s", persona1.nom);
    printf("Dirección: %s", persona1.dir);
    printf("Teléfono: %ld\n", persona1.telefono);
    printf("Fecha de nacimiento: %02d/%02d/%04d\n",
           persona1.nac.dia, persona1.nac.mes, persona1.nac.anyo);

    return 0;
}
```

#### Versión 2: función que recibe un puntero  

```c
#include <stdio.h>
#include "ficha.h"

// Declaración
void leerFichav2(struct ficha *estudiante);

int main() {
    struct ficha persona2;

    leerFichav2(&persona2);

    printf("\n--- Persona 2 ---\n");
    printf("Nombre: %s", persona2.nom);
    printf("Dirección: %s", persona2.dir);
    printf("Teléfono: %ld\n", persona2.telefono);
    printf("Fecha de nacimiento: %02d/%02d/%04d\n",
           persona2.nac.dia, persona2.nac.mes, persona2.nac.anyo);

    return 0;
}
```

> Mostrar cada versión en su propio `main` te ayuda a entender claramente cómo se pasan los datos y las diferencias en eficiencia.


## 3.3 Actualización del archivo de cabecera para incluir declaraciones de funciones  

Para que `main.c` pueda usar nuestras funciones y estructuras, debemos **declararlas en un archivo de cabecera** (`ficha.h`). Esto permite **modularidad y reutilización** del código, evitando errores por declaraciones repetidas o falta de prototipos.

> ⚠️ **Importante:**  
> No es obligatorio usar ambas funciones en un mismo programa. Según la versión que decidas usar:
> - Si eliges **leerFichav1()**, invocarás la función que **retorna la estructura**.  
> - Si eliges **leerFichav2()**, invocarás la función que **recibe un puntero** y modifica la variable directamente.  
> Esto permite que cada programa use la estrategia más conveniente según la estructura y contexto.


#### `ficha.h` actualizado

```c
#ifndef _FICHA_H
#define _FICHA_H

// Definición de la estructura fecha
struct fecha {
    int dia, mes, anyo;
};

// Definición de la estructura ficha
struct ficha {
    char nom[40];
    char dir[40];
    long telefono;
    struct fecha nac;
};

// Declaraciones de funciones para leer datos de la estructura
struct ficha leerFichav1();          // Versión que retorna la estructura
void leerFichav2(struct ficha *est); // Versión que recibe un puntero

#endif
```


## 4. Creación y definición de una función que usa una estructura

En este apartado aprenderemos a **definir una función que recibe una estructura como parámetro** y que **visualiza su contenido** por pantalla.  
De esta forma, podremos separar la lógica del programa principal (`main`) de la presentación o procesamiento de datos.


### Concepto clave

> Una función puede recibir estructuras **por valor** (recibe una copia) o **por referencia** (mediante puntero).  
> En este ejemplo, trabajaremos con el **paso por valor**, lo que significa que la función recibe **una copia** de la estructura original.  

### Prototipo de la función

```c
int visualizarFicha(struct ficha persona);
```

### Parámetro: 

Una variable de tipo `struct ficha`.

### Valor de retorno: 

Un entero (`int`), que se usará como código de éxito (0).  

### Posible implementación de la función:   

```c
int visualizarFicha(struct ficha persona)
{
    printf("Nombre: ");
    puts(persona.nom);

    printf("Dirección: ");
    puts(persona.dir);

    printf("Teléfono: %ld\n", persona.telefono);

    printf("Fecha de nacimiento (dd/mm/aaaa): ");
    printf("%d/%d/%d\n", persona.nac.dia, persona.nac.mes, persona.nac.anyo);

    return 0;
}
```

### Ejemplo de uso en `main():   

```c
int main() {
    struct ficha alumno = {
        "Lucas",
        "Calle del Cerezo 4",
        910002233,
        {1, 3, 2000}
    };

    visualizarFicha(alumno);

    return 0;
}
```

## 5. Copia de estructuras  

En C, las estructuras pueden copiarse directamente entre sí usando el operador de asignación (`=`), siempre que ambas estructuras sean del mismo tipo. Esto permite duplicar el contenido completo (campo a campo) de una estructura a otra de manera sencilla.  

<div class="alert alert-success" style="font-family: monospace;">
💡 <strong>Concepto clave</strong><br>
Cuando hacemos una asignación entre estructuras del mismo tipo, <strong>C copia todos los valores de sus miembros</strong>.<br>
Es una <strong>copia completa y por valor</strong>, no una referencia: modificar una no afecta a la otra.
</div>


### Ejemplo de código  

```c
#include "ficha.h"

int main() {
    struct ficha persona1, persona2;

    // Se leen los datos de la primera persona
    persona1 = leerFichav1();

    // Se copia persona1 en persona2
    persona2 = persona1;

    // Mostramos los datos de la copia
    visualizarFicha(persona2);

    return 0;
}
```

> ⚠️ Importante:  
> Esta copia solo es válida si las estructuras son del mismo tipo.
> Si una estructura contiene punteros, la copia será superficial: solo se copiarán las direcciones, no los datos apuntados. En esos casos, puede ser necesario realizar una copia profunda (`deep copy`) con código adicional.

## Ejemplo:   

```c
persona1.telefono = 999999999;

printf("Teléfono de persona1: %ld\n", persona1.telefono);
printf("Teléfono de persona2: %ld\n", persona2.telefono);
```   


## 6. Acceso a los campos de una estructura  

En C, una vez que tenemos una estructura declarada, podemos acceder a sus campos de dos maneras principales:

* Usando la propia variable de estructura con el operador punto (`.`)

* Usando un puntero a la estructura con el operador flecha (`->`)

<div class="alert alert-success" style="font-family: monospace;"> 
💡 <strong>RECUERDA...</strong><br> 
- Una <code>estructura</code> es un conjunto de variables bajo un mismo identificador.<br> 
- Para acceder a sus campos:<br>
    * Si tenemos la variable, usamos <code>.</code> (punto).<br>   
    * Si tenemos un puntero a la variable, usamos <code>-></code> (flecha).<br>  
</div>  

### Ejemplo:  

```c
#include <stdio.h>
#include <string.h>

// Definición de las estructuras
struct fecha {
    int dia, mes, anyo;
};

struct ficha {
    char nom[40];
    char dir[40];
    long telefono;
    struct fecha nac;
};

int main() {
    // Declaración de una estructura y un puntero
    struct ficha persona;
    struct ficha *pPersona;

    // El puntero apunta a la estructura
    pPersona = &persona;

    // Acceso a los campos usando la variable directamente
    persona.telefono = 9853333330;
    printf("Teléfono usando persona: %ld\n", persona.telefono);

    // Acceso a los campos usando el puntero a la estructura
    pPersona->telefono = 910002233;
    printf("Teléfono usando puntero pPersona: %ld\n", pPersona->telefono);

    // Acceso a otros campos
    strcpy(persona.nom, "Lucas");
    strcpy(pPersona->dir, "Calle del Cerezo 4");

    printf("Nombre: %s\n", persona.nom);
    printf("Dirección: %s\n", persona.dir);

    return 0;
}
```

## 7. Crear sinónimos de estructuras (`typedef`)  

En C, podemos crear **sinónimos (alias) para estructuras** usando `typedef`.  
Esto simplifica la sintaxis y hace el código más legible, evitando escribir `struct` cada vez que declaramos variables.

<div class="alert alert-success" style="font-family: monospace;">
💡 <strong>RECUERDA...</strong><br>
- La sintaxis general para crear un sinónimo es:<br>
<code>typedef struct tipo_estructura { tipo miembro_1; tipo miembro_2; ... tipo miembro_m; } sinonimo;</code><br>
- Después de esto, podemos declarar variables usando directamente el <code>sinonimo</code> sin poner <code>struct</code>.
</div>


### Ejemplo: crear un sinónimo para <code>struct ficha</code>

```c
#include <stdio.h>
#include <string.h>

// Definición de estructuras
struct fecha {
    int dia, mes, anyo;
};

struct ficha {
    char nom[40];
    char dir[40];
    long telefono;
    struct fecha nac;
};

// Crear un sinónimo con typedef
typedef struct ficha Ficha;

void leerFicha(Ficha *persona) {
    printf("Nombre: ");
    fgets(persona->nom, sizeof(persona->nom), stdin);

    printf("Dirección: ");
    fgets(persona->dir, sizeof(persona->dir), stdin);

    printf("Teléfono: ");
    scanf("%ld", &persona->telefono);

    printf("Fecha de nacimiento (dd mm aaaa): ");
    scanf("%d %d %d", &persona->nac.dia, &persona->nac.mes, &persona->nac.anyo);
    getchar(); // limpiar buffer
}

void visualizarFicha(Ficha persona) {
    printf("\nNombre: %s", persona.nom);
    printf("Dirección: %s", persona.dir);
    printf("Teléfono: %ld\n", persona.telefono);
    printf("Fecha de nacimiento: %02d/%02d/%04d\n", \
           persona.nac.dia, persona.nac.mes, persona.nac.anyo);
}

int main() {
    // Declarar variable usando el sinónimo
    Ficha persona;

    leerFicha(&persona);
    visualizarFicha(persona);

    return 0;
}
```

### 7.1 Actualización del archivo de cabecera para incluir declaraciones de funciones y `typedef`  

Cuando trabajamos con estructuras y funciones que las manejan, es recomendable **actualizar el archivo de cabecera** (`ficha.h`) para:

- Declarar las estructuras.
- Declarar los prototipos de funciones.
- Crear sinónimos (`typedef`) para simplificar la sintaxis.

<div class="alert alert-success" style="font-family: monospace;">
💡 <strong>RECUERDA...</strong><br>
- Mantener el archivo de cabecera <strong>bien organizado</strong> facilita su inclusión en varios archivos <code>.c</code>.<br>
- Usar <strong>include guards</strong> evita redefiniciones al incluir el mismo archivo en distintos módulos.<br>
- Declarar funciones y typedef aquí permite usar <code>Ficha</code> y las funciones sin repetir código.
</div>


### Ejemplo de `ficha.h` actualizado

## 8. Arrays de estructuras

En C, podemos manejar **colecciones de estructuras** usando **arrays**, lo que nos permite almacenar y procesar múltiples elementos del mismo tipo de manera ordenada.

<div class="alert alert-success" style="font-family: monospace;">
💡 <strong>RECUERDA...</strong><br>
- Un <code>array</code> de estructuras es simplemente un conjunto de variables de tipo <code>struct</code> bajo un mismo nombre.<br>
- Se accede a cada elemento mediante su índice: <code>estudiantes[i]</code>.<br>
- Las estructuras dentro de un array pueden copiarse individualmente o usando funciones como <code>memcpy</code> para copiar el bloque completo.
</div>


### Declaración y lectura de arrays de estructuras

```c
#include "ficha.h"
#define MAX 100

int main() {
    Ficha estudiantes[MAX]; // Array de 100 estudiantes
    int i;

    // Leer datos de cada estudiante
    for (i = 0; i < MAX; i++) {
        leerFicha(&estudiantes[i]);
    }

    // Mostrar los datos de cada estudiante
    for (i = 0; i < MAX; i++) {
        visualizarFicha(estudiantes[i]);
    }

    return 0;
}
```

### Copiar arrays de estructuras elemento por elemento  

```c
#include "ficha.h"
#define MAX 100

int main() {
    Ficha estudiantes[MAX], personas[MAX];
    int i;

    // Leer datos de los estudiantes
    for (i = 0; i < MAX; i++) {
        leerFicha(&estudiantes[i]);
    }

    // Copiar cada estructura individualmente
    for (i = 0; i < MAX; i++) {
        personas[i] = estudiantes[i];
    }

    return 0;
}
```

### Copiar arrays completos usando `memcpy`  

```c
#include "ficha.h"
#include <string.h>
#define MAX 100

int main() {
    Ficha estudiantes[MAX], personas[MAX];
    int i;

    // Leer datos de los estudiantes
    for (i = 0; i < MAX; i++) {
        leerFicha(&estudiantes[i]);
    }

    // Copiar todo el array de golpe
    memcpy(personas, estudiantes, sizeof(estudiantes));

    return 0;
}
```
