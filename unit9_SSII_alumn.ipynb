## 1. Introducci√≥n    

En C, los programas necesitan **almacenar y gestionar informaci√≥n** durante su ejecuci√≥n: resultados de c√°lculos, datos de sensores, informaci√≥n de usuario, etc. Para ello utilizan la **memoria principal**, un recurso esencial en ingenier√≠a y sistemas embebidos.

C ofrece un **acceso directo y preciso a la memoria**, lo que permite crear programas muy eficientes y cercanos al hardware. Esta potencia requiere responsabilidad: el programador debe **reservar, usar y liberar memoria** de forma correcta para evitar errores, fugas de memoria o comportamientos inesperados.

La memoria de un programa en C se puede clasificar en tres tipos:

- **Memoria autom√°tica**: la que utilizan las **variables locales** dentro de funciones. Se gestiona autom√°ticamente y desaparece al salir de la funci√≥n.  
- **Memoria est√°tica**: la que usan las **variables globales** o declaradas con `static`. Existe durante toda la ejecuci√≥n del programa.  
- **Memoria din√°mica**: se asigna **en tiempo de ejecuci√≥n** usando funciones especiales. Es la que permite crear estructuras de tama√±o variable y gestionar memoria seg√∫n las necesidades del programa.

En este tema nos centraremos en la memoria din√°mica, que permite crear estructuras de tama√±o variable y gestionar la memoria durante la ejecuci√≥n. Su uso correcto es esencial para desarrollar programas robustos y eficientes en entornos como firmware, sistemas embebidos y aplicaciones que requieren control directo sobre el hardware.

> Un mal manejo de la memoria din√°mica puede producir fugas de memoria, errores y comportamientos inesperados.

### 1.1 Memoria principal

Cuando ejecutamos un programa en C, el **sistema operativo** reserva un **espacio de memoria** para √©l, dividido en varias zonas.

#### CODE ‚Äî C√≥digo del programa

- Aqu√≠ se almacena el **c√≥digo ejecutable**.
- Suele ser de **solo lectura** para evitar cambios accidentales.
- El compilador genera el c√≥digo, y el sistema operativo lo carga en esta regi√≥n.

#### DATA ‚Äî Variables globales y literales

- Aqu√≠ viven las **variables globales**, **est√°ticas** y las **literales de cadena**.
- Estas variables existen durante toda la ejecuci√≥n del programa.
- Est√°n listas antes de que `main()` empiece.

#### HEAP ‚Äî Memoria din√°mica

- Zona donde se guarda la memoria solicitada **en tiempo de ejecuci√≥n**.
- Se usa para estructuras y arrays cuyo tama√±o se decide mientras corre el programa.
- Una mala gesti√≥n puede causar errores o fugas de memoria.

#### STACK ‚Äî Pila de ejecuci√≥n

- Almacena **variables locales**, **par√°metros de funciones** y la **direcci√≥n de retorno**.
- Crece y decrece autom√°ticamente al llamar y finalizar funciones.
- Normalmente crece hacia abajo.

#### Barrera entre STACK y HEAP

- Evita que el **stack** y el **heap** se solapen.
- Previene fallos graves durante la ejecuci√≥n.

#### Papel del compilador

- No necesitamos preocuparnos por la ubicaci√≥n exacta de las variables o funciones.
- El compilador traduce nuestro c√≥digo a instrucciones que acceden correctamente a cada dato.
- Cuando usamos una variable, se asegura de que se lea o escriba en el lugar correcto.

### 1.2 Memoria din√°mica

En C, la **memoria din√°mica** es un espacio que el programa puede pedir mientras se est√° ejecutando. Una analog√≠a: **tomar prestado un libro de la biblioteca**:  

- Pides el libro cuando lo necesitas.  
- Lo devuelves cuando terminas de usarlo.  

Si no lo devuelves, otros no pueden usarlo. En programaci√≥n, **no liberar la memoria din√°mica** genera **fugas de memoria**, provocando errores o que el programa use m√°s recursos de los necesarios.


#### Propiedades de la memoria din√°mica

- Permite guardar informaci√≥n que debe permanecer disponible aunque ciertas funciones ya hayan terminado.  
- La memoria se **solicita y libera expl√≠citamente**, dando control total al programador.  
- Est√° en el **heap**, distinto del *stack* donde est√°n las variables locales.


#### Funciones para manejar memoria din√°mica

En C se usan principalmente cuatro funciones:  

1. **`malloc` (*memory allocation*)**  
   - Reserva un bloque de memoria de un tama√±o espec√≠fico.  
   - No inicializa el contenido (los valores son indeterminados).  

2. **`calloc` (*contiguous allocation*)**  
   - Reserva memoria para varios elementos y **la inicializa a cero**.  
   - √ötil para arrays cuando queremos valores iniciales conocidos.  

3. **`realloc` (*re-allocation*)**  
   - Cambia el tama√±o de un bloque previamente reservado.  
   - Puede mover el bloque si no hay espacio contiguo suficiente.  

4. **`free`**  
   - Libera memoria previamente reservada.  
   - Evita fugas de memoria y permite reutilizar el espacio del *heap*.

### 2.1 Funciones para reservar memoria din√°mica

## 3. Reserva de memoria: tres modelos

En C existen tres **modelos principales** para usar memoria din√°mica. Cada uno organiza los datos en el **Heap** de manera diferente:

1. **Array din√°mico** ‚Äì Lista de elementos del mismo tipo, en memoria contigua. √ötil para almacenar secuencias lineales de datos, como n√∫meros o registros de un mismo tipo.  
2. **Array de punteros** ‚Äì Cada elemento apunta a otra estructura o cadena. Se usa cuando queremos manejar colecciones heterog√©neas o listas de cadenas de longitud variable.  
3. **Puntero a punteros** ‚Äì Para matrices din√°micas u otras estructuras donde cada fila puede tener distinto tama√±o. Permite organizar datos en tablas bidimensionales o m√°s complejas.


**Conceptos importantes**  

- Un **puntero** debe declararse del mismo tipo que el valor al que apunta.  
- Siempre debe **inicializarse correctamente**, ya sea con `NULL` o con la direcci√≥n de una variable existente.  
- Cada modelo permite **estructurar y acceder a los datos de manera distinta**, seg√∫n las necesidades del programa.

### 3.1 Modelo 1: Array din√°mico unidimensional

En este modelo veremos c√≥mo crear un **array lineal** usando **memoria din√°mica**, lo que permite reservar espacio para un n√∫mero de elementos que solo se conoce en tiempo de ejecuci√≥n.

Un array din√°mico funciona as√≠:

**Representaci√≥n conceptual: Array din√°mico de `float` (`pfloat`)**

```
Antes de reservar memoria:

(Stack Memory)                
+----------------+
| pfloat -------> NULL
+----------------+
| ...            |

Despu√©s de reservar memoria din√°mica para n elementos (malloc()/calloc()):

(Stack Memory)                (Dynamic Memory)
+----------------+          +----------------------------+
| pfloat ------->|  [0]     | valor: ...                 |
|                |  [1]     | valor: ...                 |
|                |  [2]     | valor: ...                 |
|                |  ...     | ...                        |
|                |  [n-1]   | valor: ...                 |
+----------------+          +----------------------------+
| ...            |          | ...                        |
```

**Notas clave:**

1. `pfloat` est√° en **stack** y apunta al bloque de memoria din√°mica en **heap**.  
2. Cada celda del array din√°mico contiene un `float`.  
3. La memoria se reserva din√°micamente con `malloc()` o `calloc()`:


#### Reserva de memoria din√°mica (`malloc()` / `calloc()`)

**`malloc(size)`**

- Reserva memoria **sin inicializar**.  
- Valores iniciales desconocidos ("basura").

```c
pfloat = (float *)malloc(sizeof(float) * n);
```

**`calloc(n, size)`**  

- Reserva memoria inicializada a cero.
- √ötil para evitar valores basura.

```c
pfloat = (float *)calloc(n, sizeof(float));
```

> **Importante**: el resto del programa es exactamente igual: lectura ‚Üí uso ‚Üí `free()`.

#### C√≥digo completo del modelo 1  

```c
#include <stdio.h>
#include <stdlib.h>

int main() {
    int i, n;
    float *pfloat;

    // ‚ë† Preguntar cu√°ntos elementos quiere el usuario
    printf("Number of elements: ");
    scanf("%d", &n);

    // ‚ë° Reservar memoria din√°mica
    //    A: malloc  ‚Üí memoria sin inicializar
    //    B: calloc ‚Üí memoria inicializada a cero
    pfloat = (float *)malloc(sizeof(float) * n);
    // pfloat = (float *)calloc(n, sizeof(float));   // ‚Üê alternativa

    // Comprobaci√≥n obligatoria
    if (pfloat == NULL) {
        printf("Error allocating memory\n");
        return 1;
    }

    /*
        Heap tras malloc:
        +-----------------------+
        | ??? | ??? | ... | ??? |
        +-----------------------+
        (valores indeterminados)
    */

    // ‚ë¢ Leer datos desde teclado
    for (i = 0; i < n; i++) {
        printf("Enter element %d: ", i);
        scanf("%f", pfloat + i);
    }

    /*
        Heap tras lectura:
        +-----------------------+
        | f0 | f1 | ... | f(n-1)|
        +-----------------------+
    */

    // ‚ë£ Mostrar valores introducidos
    printf("\nArray entered:\n");
    for (i = 0; i < n; i++) {
        printf("%f ", pfloat[i]);
    }

    // ‚ë§ Liberar la memoria reservada
    free(pfloat);

    return 0;
}
```

#### 3.1.1 Encapsulaci√≥n de la reserva de memoria

En este modelo seguimos con el **array din√°mico de una dimensi√≥n**, pero ahora encapsulamos la reserva de memoria en una funci√≥n.  
Esto permite separar la l√≥gica de asignaci√≥n de memoria del resto del programa, manteniendo el mismo flujo que vimos antes.

El concepto de memoria sigue siendo el mismo.

**Prototipo de la funci√≥n de reserva**  

```c
float* ReservaMemoria(int num);
```
* **Par√°metros:**
    * **num**: n√∫mero de elementos que tendr√° el array.
* **Valor devuelto**: puntero al bloque de memoria reservado.
    * El puntero ser√° `NULL` si no se puede reservar memoria.

**Definici√≥n de la funci√≥n:**

```c
#include <stdio.h>
#include <stdlib.h>

// Funci√≥n que encapsula la reserva de memoria para 'num' floats
float* ReservaMemoria(int num) {
    float *p = NULL;

    // ‚ë† Reserva memoria din√°mica
    p = (float*)malloc(sizeof(float) * num);
    // p = (float*)calloc(num, sizeof(float));  // alternativa: memoria inicializada a cero

    // ‚ë° Comprobaci√≥n obligatoria
    if (p == NULL) {
        printf("\nError allocating memory\n");
    }

    // ‚ë¢ Devuelve puntero al bloque reservado
    return p;
}
```

**C√≥digo de la funci√≥n `main()`:**  

```c
int main() {
    int i, n;
    float *pfloat;

    // ‚ë† Preguntar n√∫mero de elementos
    printf("Number of elements: ");
    scanf("%d", &n);

    // ‚ë° Llamar a la funci√≥n de reserva
    pfloat = ReservaMemoria(n);

    // ‚ë¢ Comprobaci√≥n del puntero
    if (pfloat != NULL) {

        // ‚ë£ Leer datos del usuario
        for (i = 0; i < n; i++) {
            printf("Enter element %d: ", i);
            scanf("%f", pfloat + i);
        }

        // ‚ë§ Mostrar valores
        printf("\nArray entered:\n");
        for (i = 0; i < n; i++) {
            printf("%f ", pfloat[i]);
        }

        // ‚ë• Liberar memoria
        free(pfloat);
    }

    return 0;
}
```

#### Funci√≥n din√°mica para crear un array de enteros

En este apartado vamos a crear una funci√≥n que genere un **array de enteros de forma din√°mica**, donde el **n√∫mero de elementos se pasa como par√°metro**.

Inicialmente (1 elemento le√≠do):

(Stack Memory)                (Dynamic Memory - Heap)
+----------------+           +----------------------------+
| pFicha --------+---------> | Ficha1                     |  [0]
|                |           |   nombre: "Vicente"        |
|                |           |   nota: 7.5                |
+----------------+           +----------------------------+
| ...            |           | ...                        |

Despu√©s de realloc (a√±adimos 2 elementos m√°s, sin inicializar):

(Stack Memory)                (Dynamic Memory - Heap)
+----------------+           +----------------------------+
| pFicha --------+---------> | Ficha1                     |  [0]
|                |           |   nombre: "Vicente"        |
|                |           |   nota: 7.5                |
|                |           +----------------------------+
|                |           | Ficha1                     |  [1]
|                |           |   nombre: ?                |
|                |           |   nota: ?                  |
|                |           +----------------------------+
|                |           | Ficha1                     |  [2]
|                |           |   nombre: ?                |
|                |           |   nota: ?                  |
|                |           +----------------------------+
| ...            |           | ...                        |
```


**Prototipo de la funci√≥n de redimensionamiento del bloque de memoria**  

```c
void* realloc(void* ptr, size_t new_size);
```
* **Par√°metros:**
    * **ptr**: puntero al bloque previamente reservado con `malloc()`, `calloc()` o `realloc()`.
        - Si ptr es `NULL`, `realloc()` se comporta como `malloc()`.
    * **new_size**: nuevo tama√±o en bytes que queremos reservar.
* **Valor devuelto**: Puntero al bloque redimensionado.
    * `NULL` si no hay memoria suficiente (en este caso el bloque original no se libera, sigue v√°lido).

>  ‚ö†Ô∏è **Aviso importante**: Siempre debemos asignar el resultado de `realloc()` a un puntero temporal para evitar perder la referencia al bloque original en caso de error.

#### Ejercicio 1 de aplicaci√≥n: Redimensionar un array de enteros

```c
#include <stdio.h>
#include <stdlib.h>

int main() {
    int *array = NULL;
    int n = 3, i;

    // ‚ë† Reserva inicial de memoria para 3 elementos
    array = (int*)malloc(sizeof(int) * n);
    if (array == NULL) {
        printf("Error allocating memory\n");
        return -1;
    }

    // ‚ë° Leer valores iniciales
    for (i = 0; i < n; i++) {
        printf("Enter element %d: ", i);
        scanf("%d", array + i);
    }

    // ‚ë¢ Redimensionar array a 5 elementos
    int *temp = (int*)realloc(array, sizeof(int) * 5);
    if (temp == NULL) {
        printf("Error reallocating memory\n");
        free(array);  // liberar bloque original. Opcional, si ya no los necesitamos, liberar
        return -1;
    }
    array = temp;  // actualizar puntero al bloque redimensionado

    // ‚ë£ Leer valores nuevos
    for (i = 3; i < 5; i++) {
        printf("Enter element %d: ", i);
        scanf("%d", array + i);
    }

    // ‚ë§ Mostrar todos los valores
    printf("\nArray completo:\n");
    for (i = 0; i < 5; i++) {
        printf("%d ", array[i]);
    }

    // ‚ë• Liberar memoria
    free(array);

    return 0;
}
```

En este ejemplo vamos a trabajar con un **array din√°mico de estructuras** `tpalabra`.  
Queremos poder **a√±adir un nuevo elemento** al array, **redimensionando din√°micamente** la memoria con `realloc`, de forma segura, y encapsulando la operaci√≥n en una funci√≥n.

**Prototipo de la funci√≥n para a√±adir un elemento:**

```c
tpalabra* ReservaMemoria(tpalabra *p, int *num);
```
* **Par√°metros:**
    * **p**: puntero al array de `tpalabra`.
    * **num**: n√∫mero de elementos actuales del array (pasado por referencia).
* **Valor devuelto**: Puntero al array redimensionado si la reserva tuvo √©xito.
    * Si la reserva falla, se devuelve el puntero original, de modo que el bloque previo sigue siendo v√°lido.

**Definici√≥n de la funci√≥n para a√±adir un elemento:**  

```c
#include <stdio.h>
#include <stdlib.h>

typedef struct {
    char palabra[50];
} tpalabra;

/*
 * ReservaMemoria:
 *   - Recibe el puntero al bloque actual y el n√∫mero de elementos que contiene.
 *   - Intenta redimensionar el bloque con realloc para a√±adir 1 elemento.
 *   - Si realloc falla, NO perdemos el puntero original.
 *   - Si tiene √©xito, incrementa el contador y devuelve el nuevo puntero.
 */
tpalabra* ReservaMemoria(tpalabra *p, int *num) {
    tpalabra *p_aux = NULL;

    // Intento de redimensionar el bloque. realloc conserva el contenido.
    p_aux = realloc(p, sizeof(tpalabra) * (*num + 1));

    if (p_aux == NULL) {
        /*
         * IMPORTANTE:
         * Si realloc falla:
         *    - NO debemos perder el puntero original.
         *    - Por eso devolvemos p (el original).
         *    - num NO debe aumentar.
         */
        printf("Error allocating memory.\n");
        return p;
    }

    /*
     * Si realloc tiene √©xito:
     *    - p_aux apunta al BLOQUE ACTUALIZADO.
     *    - p queda obsoleto, debemos usar p_aux.
     *    - Incrementamos el n√∫mero de elementos.
     */
    (*num)++;
    return p_aux;
}
```

```c
#include <stdio.h>
#include <stdlib.h>
#include <string.h>

typedef struct {
    char palabra[50];
} tpalabra;


tpalabra* ReservaMemoria(tpalabra *p, int *num);   // Prototipo de redimensionar memoria

int main() {
    tpalabra *pFicha = NULL;
    int num = 1;

    /*
     * ‚ë† Reserva inicial de memoria para UN elemento
     *    malloc retorna NULL si falla, lo que se debe comprobar.
     */
    pFicha = malloc(sizeof(tpalabra) * num);
    if (pFicha == NULL) {
        printf("\nError allocating memory\n");
        return -1;
    }

    /*
     * ‚ë° A√±adir un nuevo elemento mediante realloc (a trav√©s de la funci√≥n)
     *    Si falla, pFicha sigue siendo v√°lido (aunque no ampliado).
     */
    pFicha = ReservaMemoria(pFicha, &num);

    /*
     * ‚ë¢ Lectura segura de los elementos
     *    Se leen "num" estructuras gracias a realloc.
     */
    if (pFicha != NULL) {
        for (int i = 0; i < num; i++) {
            printf("Palabra %d: ", i);

            /*
             * scanf("%49s") evita desbordamientos.
             * Nunca usar scanf("%s") sin l√≠mite.
             */
            scanf("%49s", pFicha[i].palabra);
        }

        // ‚ë£ Mostrar todos los elementos
        printf("\nArray completo:\n");
        for (int i = 0; i < num; i++) {
            printf("%s\n", pFicha[i].palabra);
        }

        // ‚ë§ Liberar memoria
        /*
         * Muy importante:
         *    - S√≥lo se hace free UNA vez al puntero devuelto por realloc/malloc.
         *    - Nunca liberar p_aux o p antiguo.
         */
        free(pFicha);
        pFicha = NULL;
    }

    return 0;
}

tpalabra* ReservaMemoria(tpalabra *p, int *num) {
    tpalabra *p_aux = NULL;

    p_aux = realloc(p, sizeof(tpalabra) * (*num + 1));

    if (p_aux == NULL) {
        
        printf("Error allocating memory.\n");
        return p;
    }

    
    (*num)++;
    return p_aux;
}
```

En este apartado trabajaremos con arrays din√°micos de estructuras, donde cada estructura almacena informaci√≥n de una persona.
Usaremos `malloc()` para reservar memoria para los campos internos (por ejemplo, para el nombre) y `realloc()` para redimensionar din√°micamente el array de estructuras seg√∫n crezca la informaci√≥n.

**Declaraci√≥n de la estructura**

```c
typedef struct card2 {
    char *pnombre;
    float nota;
} Ficha2;
```

**Representaci√≥n conceptual: Arrays din√°micos de estructuras `Ficha2` con `malloc()`**

Ejemplo: reservamos memoria din√°mica con `malloc()` s√≥lo para el nombre de una persona cuya informaci√≥n es almacenada en la variable `ficha` de tipo `Ficha2`, que est√° en *stack*.

```
(Stack Memory)                    (Dynamic Memory - Heap)

+-------------------+                 
| ...               |                 
+-------------------+           +---+---+---+---+---+----+
| ficha             |           |'M'|'a'|'r'|'i'|'o'|'\0'|
| pnombre ----------+---------> +---+---+---+---+---+----+
| nota = 7.5        |                 
+-------------------+                 
| ...               |                 
+-------------------+                 
```

**Representaci√≥n conceptual: Arrays din√°micos de estructuras `Ficha2` con `realloc()`**

```
Antes de AnyadirUnaStruct() (n = 2):

(Stack)                     (Heap)

+-----------+
|   ...     |
+-----------+
| pFichas   |----->   +------------+      +--------------+
+-----------+         | pnombre ---+----> | A | n | a |\0|
|   ...     |         | nota=5.0   |      +--------------+
+-----------+         +------------+
                      | pnombre ---+----> +--------------+
                      | nota=7.5   |      | E | v | a |\0|
                      +------------+      +--------------+
```
> Donde `pFichas` es un puntero en *stack* al array din√°mico; cada `pnombre` apunta a su propia memoria din√°mica; y `nota` forma parte de cada estructura.

```
Despu√©s de AnyadirUnaStruct() con realloc() (n = 3):

+-----------+
|   ...     |
+-----------+
| pFichas   |----->   +------------+      +--------------+
+-----------+         | pnombre ---+----> | A | n | a |\0|
|   ...     |         | nota=5.0   |      +--------------+
+-----------+         +------------+
                      | pnombre ---+----> +--------------+
                      | nota=7.5   |      | E | v | a |\0|
                      +------------+      +--------------+
                      | pnombre ---+----> +------------------+
                      | nota=6.3   |      | J | o | s | e |\0|
                      +------------+      +------------------+
```

#### Funci√≥n para a√±adir una estructura al array din√°mico (`realloc`)

**Prototipo de la funci√≥n para a√±adir una estructura:**

```c
int AnyadirUnaStruct(Ficha2 **pp, int num);
```

* **Par√°metros:**
    * **pp**: puntero al bloque de memoria del array, pasado **por referencia**.
    * **num**: n√∫mero de elementos existentes en el array.
* **Valor devuelto**: Nuevo n√∫mero total de elementos despu√©s de a√±adir uno.


**Definici√≥n de la funci√≥n para a√±adir una estructura:**  

```c
int AnyadirUnaStruct(Ficha2 **pp, int num) {
    char string1[MAX_CAD];
    Ficha2 *pAux;

    printf("\nNombre? ");
    LeeCadena(string1, MAX_CAD);

    // Redimensionar array de estructuras
    pAux = (Ficha2 *)realloc(*pp, sizeof(Ficha2) * (num + 1));
    if (pAux != NULL) {
        if (pAux != *pp) *pp = pAux;

        // Reservar memoria para el nombre de la nueva estructura
        pAux[num].pnombre = (char *)malloc(sizeof(char) * (strlen(string1) + 1));
        strcpy(pAux[num].pnombre, string1);

        // Leer nota
        printf("Nota? ");
        scanf("%f", &pAux[num].nota);

        num++; // Incrementar contador
    } else {
        printf("Error allocating memory.\n");
    }

    return num;
}
```

**C√≥digo ejemplo de la funci√≥n `main()` usando la funci√≥n para a√±adir estructuras:**  

```c
#include <stdio.h>
#include <stdlib.h>
#include "Ficha2.h"   // Suponiendo que la estructura Ficha2 y AnyadirUnaStruct est√°n declaradas aqu√≠

int main() {
    int n = 0;
    Ficha2* pFichas = NULL;
    int i;
    char opcion;

    do {
        // A√±adir una nueva estructura al array din√°mico
        n = AnyadirUnaStruct(&pFichas, n);

        // Preguntar al usuario si quiere a√±adir otra
        printf("¬øDesea a√±adir otra persona? (s/n): ");
        scanf(" %c", &opcion);
        getchar(); // limpiar el buffer
    } while (opcion == 's' || opcion == 'S');

    // Mostrar los datos introducidos
    printf("\nListado de personas:\n");
    for (i = 0; i < n; i++) {
        printf("%d. Nombre: %s, Nota: %.2f\n", i + 1, pFichas[i].pnombre, pFichas[i].nota);
    }

    // Liberar memoria de cada cadena
    for (i = 0; i < n; i++) {
        free(pFichas[i].pnombre);
    }

    // Liberar memoria del array din√°mico
    free(pFichas);

    return 0;
}
```
